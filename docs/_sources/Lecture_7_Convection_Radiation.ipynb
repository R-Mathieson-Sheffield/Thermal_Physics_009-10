## Convection

Convection is the mechanism by which heat is transferred through fluids, where a fluid may be either a liquid or a gas. 

The Met Office defines convection as “*the transfer of heat by the circulation or movement of heated parts of a liquid or gas” or “the movement within a liquid or a gas driven by differences in temperature*”. You may be thankful to know that you will not be asked to perform any calculations for a convection heat transfer and this is because it is very complicated to do so. 

Convection can be modelled, but we find that it is a chaotic system, this means that making very small changes to the starting parameters of our model will result in drastic changes in the system as it evolves over time. 
If we consider the MET office again, when modelling weather patterns and extrapolating forwards to make predictions, they must take into consideration chaotic systems such as convection and turbulence. 

To help aid these calculations they use a supercomputer called the cray XC40, which is capable of making 14,000 trillion arithmetic calculations per second! This would be the same as every person on the earth performing over 2 million calculations per second.Thanks to these computers, the MET uses 215 billion weather measurements from across the globe every day to act as the starting point for their atmospheric models for weather predictions. 

**So what are the actual mechanics driving convection currents?**

1. Well there is a source of heat that supplies energy to one particular spot, for example, if you imagine we have a large water bath and place a bunsen burner in the middle of it, that spot in the middle is going to receive the majority of the heat supplied. 



2. This means that only the water in that spot is going to be heated significantly compared to the rest of the water in this large water bath. If we provide heat energy we are going to increase the kinetic energy of the molecules in this region which in turn is going to increase the amount of volume that they occupy. As the amount of water being heated has not changed, but the volume has increased, the density of this heated water is reduced. 


3. Denser materials sink and less dense materials rise, this means that this heated water rises and the volume it occupies is replaced with cooler water that can be then be heated.



4. As the hot water rises, it spreads out and dissipates its energy with the rest of the fluid causing it to cool down and again become more dense causing it to begin to sink. This constant rising and falling of the hot and cooling fluid is what we describe as a convection current.

### Convection Current Examples

Here we can see a model of the convection currents in the earth's mantle that nicely demonstrates this localised heating, rising and falling.

![](Figures/EarthConvection.png)

Another example of a convection current that you may be familiar with is a land/sea breeze, which can be created under the right conditions:

![](Figures/SeaBreeze.png)

- At night time, the land cools relatively quickly, but the warmer water heats the air above it, increasing the temperature of the air above the sea relative to the land. This causes the warmer less dense air to rise and be replaced by the cooler air from above the land creating a land breeze. 
 

- The opposite is true in the daytime. The land is heated up by the sun much more rapidly than the sea, causing the air above the land to heat up more quickly relative to the air above the sea. This means that the less dense air above the land rises to be replaced by the denser cooler air from the sea. This time the breeze is blowing towards the land from the sea and this is what we refer to as a sea breeze.

![](Figures/WeatherFronts.png)

Convection currents are very important when modelling climate and weather systems and they can be seen to operate on a global scale.  As the sun heats the surface of the earth, the air above the surface is heated and rises, and under certain conditions this air can continue to rise to form different types of clouds, which obviously impact the weather. 

Furthermore, there is a process known as the ocean conveyor current, which is a water current that carries warmer water from the equator of the planet to the poles. This process is impacted by convection due to the ocean temperature and the salinity (salt concentration) of the water. Warm waters near the equator are pushed towards the poles by strong wind patterns, and are carried to the higher latitudes where they cool via interaction with the colder winds. The water also evaporates during this movement, which in turn increases the salt concentration of the water. 

This increase in salinity and decrease in temperature causes the water to sink through a convection process. This is vital for driving ocean currents as it supplies water to deep water currents which then transport the water back to the tropical regions. These systems are very complicated and are made even more difficult when considering the coriolis force caused by the Earth’s rotation. 

If the earth did not rotate, the atmosphere would simply cycle from the high pressure regions at the poles to the low pressure regions at the equator and back again, however due to the earth's rotation, the air is deflected. If the air is travelling from the north pole to the equator, it is deflected west, whereas if it is travelling from the equator to the north pole it is deflected to the east. The opposite is true for the south pole. Clearly this just adds another level of complexity that makes modelling large climate systems more difficult.

## Radiation

Simply put, all objects emit electromagnetic radiation and the peak wavelength they emit is dependent on the temperature of the object.

You will recall that during the pandemic it was common to use non-contact thermometers to determine the body temperature of a person to gauge if they have a fever. These thermometers focus the infrared radiation that we radiate onto a sensor in order to determine skin temperature. We are also aware that we do not need a thermal imaging camera to know that what we have said here is true, as we heat an object to a higher and higher temperature, eventually the electromagnetic wave it emits will fall within the visible region of the electromagnetic spectrum. For example, the image above shows a blacksmith working on a glowing red hot piece of metal. Things become red hot at around $800$˚C and white hot at around $3000˚$C.

### Blackbody Radiation

Now, when we want to perform calculations to determine the temperature of an object from the wavelength of radiation it emits, we do so assuming that it is a *blackbody radiator*. 

A blackbody radiator is a hypothetical that object is capable of emitting and absorbing all wavelengths of electromagnetic radiation. In reality, there are no ideal black bodies, but there are some examples that are good approximations. 

The best example is the cosmic microwave background (CMB), more commonly referred to as “the afterglow of the big bang”. If you ever watched an analogue TV (and it scares me slightly that you might be young enough to not have!) that was not tuned to a channel you would have seen TV static - this is the jumping black and white dots that flicker on the screen - part of this static pattern may be attributed to the cosmic microwave background. ```

Now we have established what a blackbody radiator is, we can begin to investigate the relationship between the wavelength of light that an object emits and its temperature. 

---
## The Planck Curve

An object at a given temperature will radiate electromagnetic radiation, however it does not emit a single wavelength of electromagnetic radiation, instead it radiates a distribution of wavelengths, the intensities of which vary as a function of temperature.

We can plot the *spectral radiance* as a function of wavelength for an object at various temperatures to see how this changes. Spectral radiance is simply “the radiance of a surface per unit frequency or wavelength” and has units of kW$\,$sr$^{-1}\,$m$^{-2}\,$nm$^{-1}$ if measured as a function of wavelength. In essence it is a  measure of intensity. The diagram below is known as a Planck curve and illustrates how the spectral radiance varies as a function of temperature:

![](Figures/PlankCurve.png)

What we notice is that as the temperature of a body increases, the peak wavelength of the light it emits increases in intensity and is shifted towards the lower wavelengths (e.g. higher frequencies/energies). 
This is a very important relationship to remember and one that can only be explained based on the teachings of quantum mechanics. The black line on this plot shows the prediction made using Classical theory (e.g. not quantum theory) - we see that as the wavelength emitted by the object gets smaller and smaller into the ultraviolet region, the intensity of its spectral radiance increases exponentially to infinity! 

This means that if we were to integrate under the curve to find the total radiated intensity, it would also be equal to infinity - this is clearly not possible. Lord Rayleigh who derived this classical model referred to this as the "ultraviolet catastrophe".

Plancks' model based on quantum mechanics was not favoured at the time even by himself - in a letter to his friend he called it “an act of desperation” into which he was forced because “a theoretical explanation had to be found at any cost, regardless of the price”. Only five years later with the  Einstein's explanation of the photoelectric effect amongst other evidence, did the validity of quantum theory become accepted. 

---
### Wien's Displacement Law

The temperature, $T$, of an object is related to the peak wavelength of light it emits, $\lambda_{peak}$, by Wien's Displacement Law:

$ \lambda_{peak} = \frac{b}{T}, $

where $b$ is Wiens displacement constant, which is equal to $2.8977721(26) \times 10^{-3}\,$m$\,$K$^{-1}$. The peak wavelength therefore has units of $m$ and the temperature is measured in Kelvin.

---
### Stefan's Law

There is a second law that we should learn when considering blackbody radiators and their temperature dependencies - this is Stefan’s Law. Stefan’s Law tells us the total power, $P$ emitted by a blackbody radiator with a given surface area, $A$, and temperature, $T$:

$P = A \sigma T^4, $

where $\sigma$ is the Stefan-Boltzmann constant and has a value of $5.67 \times 10^8\,$W$\,$m$^{-2}\,$K$^{-4}$. This relationship is used routinely in astronomy to gain information about stars, for example, the stage at which the star is in its life cycle. We can do this using a Hertzsprung-Russel Diagram (shown below).

![](Figures/RHDiagram.png)

The total power of a star is simply the amount of energy it emits in a given unit of time (recall that $P = \frac{dE}{dt}$). In astronomy this is often just called the luminosity of the star.  Using Stefan’s law we are able to determine the surface temperature of a star from its luminosity. 

We can see from this diagram that if we plot a star's luminosity (in this case measured relative to the sun) as a function of its surface temperature, the stars fall within different clusters depending on the type of star they are. For example, white dwarves, main sequence stars, red or blue giants or supergiants. 

```{warning}
The next few paragraphs of notes outline the life-cycle of a star and are suplementary to this course. They are only included for reference if you are interested. However, you should still be comfortable using Stefans Law and Wien's Displacement Law.
```

The middle strip of this plot shows main sequence stars such as the sun. These are the most numerous stars in the universe and maintain themselves by fusing hydrogen into helium in their core. This generates a radiation pressure that exerts and outwards in order to counteract the gravitational collapse of the star inward. 

Once almost all of the hydrogen has been used up, the fusion pressure in the core drops and the core of the star collapses under its gravitational forces, this increases the temperature of the core which allows hydrogen in a shell surrounding the core to begin to fuse, which causes the star to expand. As the core continues to collapse and become more and more dense, the outer shells of the star expand further and further until the star becomes a red giant.  

Low mass main sequence stars are unable to become red giants and instead will continue to fuse hydrogen for up to a trillion years - the current age of the universe is 13.7 billion years for reference! If the star is sufficiently heavy (above 8-10 solar masses) once the core of the star has fused into helium, the gravitational forces acting on the core are sufficient to fuse heavier elements until eventually the core is formed of iron (no more fusion can take place at this point as we will learn later this semester). At the end of the lifetime of the star, these super giants will go supernova and depending on their mass produce either a neutron star or a blackhole. Stars with masses less than 10 solar masses cannot go supernova and instead they just slowly lose their outer layers leaving behind only the core of the star - this is known as a white dwarf star.

### Question 1

A hot object emits light with a peak wavelength of $0.5470\,\mu m$. What is its blackbody temperature?

```{admonition} Q1 Solution
:class: dropdown

To answer this question we are simply going to use Wien's displacement law:

$ \lambda_{peak} = \frac{b}{T},$

$ T = \frac{b}{\lambda_{peak}}, $

$ T = \frac{2.898 \times 10^{-3}}{0.547 \times 10^{-6}}, $

$ T = 5298\, K.$

```
---
### Question 2

The sun has a diameter of $1.4 \times 10^9\,$m and a surface temperature of $5800\,$K. Calculate the power it radiates into space.

```{admonition} Q2 Solution
:class: dropdown

To answer this question we must use Stefan's law:

$P = A \sigma T^4, $

as $A$ is the surface area of the sun, we must write an expression for this in terms of the diameter provided:

$ A = 4\pi r^2, $

$ A = 4\pi \left(\frac{d}{2}\right)^2, $

$ A = 4\pi \frac{d^2}{4}. $

$ A = \pi d^2. $

Now we sub in the expression for the surface area into Stefan's law:

$P =  \pi d^2 \sigma T^4, $

$ P = \pi \times (1.4 \times 10^9)^2 \times 5.67\times 10^{-8} \times 5800^4$

$ P = 3.95 \times 10^{26}\, W$

$ P = 4.0 \times 10^{26}\, W$

```

### Question 3

Solar radiation falls onto the surface of the Earth at an intensity of $1.37\,$kW$\,$m$^{-2}$ junder normal incidence. Estimate the temperature of the Earth assuming it radiates back into space as much power as it recieves. The radius of the Earth is $6.4\times 10^6\,$m. 

```{admonition} Q3 Solution
:class: dropdown

To answer this question, we will make the assumption that the earth behaves as a perfect blackbody. As the Sun’s radiation is not normal to the surface of the earth (except near the equator), the area of the Earth used in this calculation will be the cross sectional area, $A_c$, rather than half the surface area:

$A_c = \pi r^2$

Looking at the units of the solar radiation intensity, it is clear that if multiplied by an area we will get the corresponding amount of power incident upon the earth:

$P = I A$

$P = I \pi r^2.$

The question also states that the same amount of power is radiated back into space, therefore we can equate the power above to the power given using Stefan's law:

$P = A \sigma T^4, $

however the Earth will radiate this energy back into space in all direction, therefore the area we use in this equation is actually the surface area of the Earth:

$P = (4 \pi r^2) \sigma T^4, $

therefore:

$I \pi r^2 = (4 \pi r^2) \sigma T^4,$

$ I = 4 \sigma T^4,$

$ T = ^4\sqrt{\frac{I}{4\sigma}}, $

$ T = ^4\sqrt{\frac{1370}{4\times 5.67\times 10^{-8}}}, $

$ T = 278.8\, K. $

This is around %6%˚C, which seems like a reasonable estimation. 

```
