## Heat, Work and Internal Energy

In the lectures so far we have discussed internal energy and heat, but now we are going to formalise these definitions further and see how they relate to the first law of thermodynamics. 

The first thing we need to define is the internal energy of a system:

```{admonition} Internal energy of a system
:class: note
The sum of the kinetic energy of all the constituent particles in a system plus all of the potential energies of the interactions between them. This is the total amount of energy contained within the body (or system).
```
- The kinetic energy of the system moving as a whole **does not** contribute to the internal energy of the system, nor does the potential energy of the system as a whole due to external forces. 
- Potential energy is represented by the symbol $U$.

The internal energy of a system can be increased in one of two ways: 

1. Supplying a quantity of heat, $Q$, to the system will increase the internal energy by the the amount of heat supplied:
```{math}
\Delta U = \Delta Q
```

2. Doing work, $W$, on the system, for example, compressing a gas in a system by doing mechanical work upon it, will also increase internal energy by an equivlent amount:
```{math}
\Delta U = \Delta W
```

In either case, the internal energy would increase by the same amount of energy as that done by doing the work or supplying the heat. This then leads us on to the formal definition of the first law of thermodynamics...

### First Law of Thermodynamics

The first law of thermodnyamics simply states that the **change** in internal energy of a system is equal to the heat added **to** the system plus the work done **on** the system. This can be expressed mathematically as:

```{math}
\Delta U = Q + W
```
This is simply a statement of conservation of energy i.e. energy cannot be created or destroyed, only converted from one form to another. 


````{warning}
A key feature that we must be aware of is the notation used. In our definition for the first law of thermodynamics we have said $W$ is positive. This is because we define $W$ as the work done **ON** the system, whereas some other texts will define this as the work done **BY** the system, which leads to the relationship:

```{math} 
\Delta U = Q - W
```

One way to think of this is to consider gas in a piston. If the system (the gas) is doing work, for example by expanding against its surroundings, energy is leaving the system and the internal energy will decrease (assuming that no heat was supplied to the system). 

In these lectures, we will only use the expression outlined outside of this warning. When you study thermodynamics on more depth, you will see the important of the first law of thermodynamics in relation to processes such as engines.
````

## Heat Capacity ($C$)

When we consider the amount of heat added to a system to increase its temperature by a given amount, we will need to consider the heat capacity of the material in question. 

The heat capacity is simply the amount of heat energy required to raise the temperature of a body by $1\text{ K}$:

```{math}
C = \frac{\Delta Q}{\Delta T}
```

Accordingly it has units of $\text{J K}^{-1}$. For example, if I had a block of glass, and I wanted to increase the temperature by 10 degrees, I would need to supply it with ten times its heat capacity worth of energy. 

In reality, it is actually fairly uncommon for us to use the general heat capacity when performing calculations, and this is because it is dependent on both the size, mass and materials of the system and as such must be determined for every system that we are performing calculations for. This is clearly not convenient.

### Specific Heat Capacity ($c$) 


It us much more common for us to perform calculations using the *specific* heat capacity, which take into account the amount of material that heat is being supplied to when calculating how much energy is required to change its temperature. 

For example, the specific heat capacity is the amount of energy required, per unit mass, to increase the temperature of a body by 1K. It is therefore related to the heat capacity via:

```{math}
:label: eq_specificheat
c = \frac{C}{M} = \frac{\Delta Q}{m \Delta T}
```

Accordingly it has units of $\text{J kg}^{-1}\text{ K}^{-1}$. Each material has its own specific heat capacity as shown below:

Note how large the specific heat capacity of water is - this means it can store a lot of heat energy!

```````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

specificHeatCapacity = [230, 390, 480, 750, 900, 920, 2200, 4200]
materialDensity = [10490, 8960, 7850, 1650, 2712, 2300, 917, 998]
material = ["  Silver", "  Copper", "  Steel", "  Brick", "  Aluminium", "  Concrete", "  Polythene", "  Water (Liquid)"]

plt.figure(figsize=(10,10))
plt.plot(materialDensity, specificHeatCapacity, marker = 'o', linestyle = ' ', color = 'b')
plt.xlabel(r'Density [$\:kg \:m^{-3}\: $]')
plt.ylabel(r'Specific Heat Capacity [$\:J \: kg^{-1} \: K^{-1}\:$]')
plt.grid()

for i, txt in enumerate(material):
    plt.annotate(txt, (materialDensity[i], specificHeatCapacity[i]))

plt.show()
```````

If we plot these specific heat capacities as a function of the material density we see a rough trend emerging - as the density of the material increases, the specific heat capacity of the material decreases. In a few topics time we will look at why this is. 

Now we have an understanding of what specific heat capacity is, we can use it to perform calculations and determine how much heat energy we must supply to or remove from an object with a given mass to change its temperature by a given amount. Rearranging the specific heat capacity equation (equation {eq}`eq_specificheat`) tells us how much heat energy is required to raise a mass, $m$, by a temperature, $\Delta T$, providing we know the specific heat capacity of the material under consideration:

```{math}
c &= \frac{\Delta Q}{c \Delta m}\\
\therefore \Delta Q &= m c \Delta T
```
This provides us with the link between the heat energy required, $\Delta Q$, and the temperature change, $\Delta T$, we will observe for a body of mass $m$.


```{admonition} A Note on Notation
Often you will see $\theta$ used in place of $T$. This is simply to note that the difference in temperatures has been measured in Celsius rather than Kelvin. 

This is immaterial as the two temperature scales increase in the same increments, e.g. if we increase the temperature of a body by $80\,^\circ\text{C}$, we have also increased its temperature by $80\text{ K}$. Therefore in our calculations it makes no matter which one we use, so long as we are looking at a change in temperature. 

Having said that, you may wish to consistently use Kelvin in your calculations as for other calculations you will perform in this unit (for example when we are working with ideal gases), the equations we use assume we are working in Kelvin. 
```

### Molar Heat Capacity (C$_{mol}$)

An alternative to the specific heat capacity is the **molar** heat capacity, which is simply the amount of energy required to increase the temperature of one mole of a substance by 1K. 

A quick reminder that "molar" simply means one mole of something, which is equal to $6.022141429 \times 10^{23}$ elementary entities (atoms or molecules) of a given substance. Therefore, one mole of a substance A will have a different mass to one mole of substance B, but they will contain the same number of elementary entities. 

The number of moles is often denoted using the letter $n$ (as opposed to the number of atoms / molecules, $N$) and as such the molar heat capacity, $C_{mol}$, can be calculated as:

```{math}
C_{mol} = \frac{\Delta Q}{n \Delta T}
```

## Changes of State and Phase Transitions

So far we have only considered supplying/removing energy to/from materials to change their temperatures, but we have not considered any phase changes. Lots of substances can display different states or phases and the physical properties of these states may differ despite the chemical substance remaining the same. 

The diagram above was shown in [Topic 3](TPS_Topic3_RevisitingFixedPoints) and illustrates the fact that phase changes are dependent on pressure and temperature in addition to the material in question, e.g. the phase diagram from water will obviously differ for oxygen. 

An important question that we need to consider is how do we determine how much energy we must supply to a material for a phase change to occur? For this we must introduce a new concept, the latent heat. 

**The specific latent heat, $L$, is the amount of energy required to change the phase of a unit mass of material.** 

One key thing to note is that phase changes occur at a constant temperature. This is to say that if we are supplying energy to a block of ice at -10˚C, initially the temperature will increase as we supply heat, but when it gets to 0˚C and begins to melt, as we continue to supply heat the temperature of the melting ice will remain the same as the energy supplied is being used to change the **state** and not the **temperature**. Then as we continue to supply energy, the resulting water will again begin to increase in temperature. 

For a given mass of a material with a latent heat of $L$, the amount of energy required to change the phase is simply equal to:

```{math}
Q = mL_x
```
where $L_x$ would be $L_f$ for fusion and $L_v$ for vaporisation. The latent heat differs for different phase transitions, for example fusion (liquid to solid) and vaporisation (liquid to gas) have different latent heats associated with them. Some materials sublimate - sublimation is when a solid turns into a gas directly without the liquid phase between. A common example of this is dry ice (solid $\text{CO}_2$).

As there is no temperature change associated with a change of phase, the unit for the specific latent heat is simply $\text{J kg}^{-1}$.

The final thing to note is that phase changes are reversible! Just as we can melt something we can freeze something! The only difference is that we are either supplying energy to or taking energy from the material, thus the internal energy of the system will either increase or decrease.

``````{admonition} Question 4.1 - Heating a metal block
`````{tab-set}
````{tab-item} Question
An aluminium block, mass $0.50\text{ kg}$, is heated by a $40\text{ W}$ electrical heater embedded in it. In $56.3$ seconds, the temperature of the block rose from $15.0\,^\circ\text{C}$ to $20.0\,^\circ\text{C}$.  
Calculate the specific heat capacity of aluminium, ignoring heat losses.
````

````{tab-item} Hint
Remember the definition of specific heat capacity. It's the amount of heat energy needed to raise the temperature of 1 kg of a substance by 1 Kelvin. 

Think about power and energy: Power is the rate at which energy is transferred. Use the formula Power (P) = Energy (E) / Time (t) to find the total energy delivered by the heater.

Break down the problem into stages. First, calculate the total heat **energy** supplied by the heater. Then find the specific heat capacity based on the change in temperature.
````

````{tab-item} Solution
Use the definition of the specific heat capacity which is the amount of energy it takes to increase the temperature of a substance by one kelvin per unit mass:

```{math}
c = \frac{\Delta Q}{m \Delta T}
```

The change in temperature, $\Delta T$ is simply $5\,^\circ\text{C} = 5\text{ K}$, and the amount of heat energy supplied can be calculated by recalling that power is simply equal to the rate at which work is done:

```{math}
P &= \frac{\Delta E}{\Delta t}\\
\therefore \Delta E &= P \Delta t
```

and so
```{math}
c &= \frac{P \Delta t}{m \Delta T}\\
    &= \frac{40 \times 56.3}{0.5 \times 5}\\
    &= 901\text{ J kg}^{-1}\text{ K}^{-1}
```
``````

`````{admonition} Question 4.2 - Heating water in a kettle
```{tab-item} Question
A kettle contains $500\,$ml of water at $20$°C. If the water is heated to $353.15\,$K in a time of 1 minute and 35 seconds.  
a) Determine how much energy the kettle must supply to the water. The specific heat capacity of water is $4200\text{ J kg}^{-1}\text{ K}^{-1}$.

```

```{tab-item} Hint
Find the heat energy $Q$ needed to heat the water by some temperature change $\Delta T$.  
Remember to convert units if necessary (e.g., mL to kg, $^\circ\text{C}$ to Kelvin).

```

```{tab-item} Solution
**Part a**  
To answer this questions, we are simply going to use the equation $Q = mc \Delta T$, all we need to do is convert the units provided in the question:

- $20\,^\circ\text{C} = 293.15\text{ K}$
- $500\text{ ml} = 500\text{ g}$[^1],

therefore, the total amount of heat energy supplied is:

\begin{align*}
Q &= mc \Delta T\\
    &= 0.5 \times 4200 \times (353.15 - 293.15)
    &= 126000\text{ J}
\end{align*}


**Part b**  
To answer this question we are again going to consider the relationship between time, energy and power. We are also going to assume that the transfer of energy from the kettle to the water was a 100% efficient process i.e. no heat losses:

- 1 minute and 35 seconds $= 95\text{ s}$

\begin{align*}
\therefore P &= \frac{\Delta E}{\Delta t}\\
    &= \frac{126000}{95}\\
    &= 1326 \text{ W}
\end{align*}

[^1]: If you did not know this, 1 cubic centimetre of water weighs $1\text{ g}$ and occupies $1\text{ ml}$ of volume.
`````

`````{admonition} Question 4.3 - Heating, melting and boiling ice
````{tab-set}
```{tab-item} Question
A $1.000\text{ kg}$ block of ice is heated from $-12.00\,^\circ\text{C}$ until it melts into water, before subsequently boiling away completely. 
a) How much energy is required to raise the temperature of the ice block to $0.000\,^\circ\text{C}$. The specific heat capacity of ice is $2108\text{ J kg}^{-1}\text{ K}^{-1}$.
b) How much energy is required to melt the ice block completely?
c) How much energy is required to increase the temperature of the resulting water to its boiling point ($100.0\,^\circ\text{C}$)? The specific heat capacity of water is $4200\text{ J kg}^{-1}\text{ K}^{-1}$.
d) How much energy is required to convert all of the water into steam? The latent heat of vaporisation is $2258\times10^3\text{ J kg}^{-1}$.
e) How long would it take a $1500\text{ W}$ heater to convert the initial block of ice completely into steam?
```

```{tab-item} Hint
Break down the problem: This question involves multiple stages: heating the ice, melting the ice, heating the water, and boiling the water.

Use the correct formulas: $Q = mc\Delta T$ and $Q = mL$.

Calculate the energy for each stage separately, and then find the total energy.

To calculate the time remember the link between power and energy involves time.
```

```{tab-item} Solution
**Part a**  
Use the equation $Q = mc\Delta T$ to determine the energy requried to increase the temperature of the ice block from $-12\,^\circ\text{C}$ to $0\,^\circ\text{C}$, which is a change in temperature of $12.00\text{ K}$:
\begin{align*}
Q &= mc \Delta T\\
    &= 1 \times 2108 \times 12\\
    &= 25300\text{ J}
\end{align*}

**Part b**
This question is about a change of phase, specifically fusion (solid to liquid) meaning that we are going to use the specific latent heat equation:
\begin{align*}
Q &= mL_f\\
    &= 1 \times 334\times10^3\\
    &= 334000\text{ J}
\end{align*}

**Part c**
This question is just a repeat of a) but this time we must use the specific heat capacity of water rather than ice and we have a temperature difference of $110\text{ K}$:
\begin{align*}
Q &= mc \Delta T\\
    &= 1 \times 4200 \times 100\\
    &= 420000\text{ J}
\end{align*}

**Part d**
This question is just a repeat of b) but this time we must use the latent heat for vaporisation rather than fusion:
\begin{align*}
Q &= mL_v\\
    &= 1 \times 2258\times10^3\\
    &= 2258000\text{ J}
\end{align*}

**Part e**
First we need to total all of the heat energy supplied for the associated phase transitions and temperature changes for:
- heating the ice
- melting the ice to water
- heating the water
- boiling the water to steam. 
Thankfully we've just calculated all of these!

\begin{align*}
E &= 25300 + 334000 + 420000 + 2258000\\
&= 3014530 \text{ J}
\end{align*}

Now we rearrange the equation for power:
\begin{align*}
P &= \frac{\Delta E}{\Delta t}\\
\therefore \Delta t &= \frac{\Delta E}{P}\\
    &= \frac{\Delta 3014530}{1500}\\
    &= 2010\text{ s}
\end{align*}
Note that this is the change in time, but we are defining our starting time ($t=0$) as the time we start heating up the ice. In this case $\Delta t = t - 0 = t$.
`````
