## Heat, Work and Internal Energy

In the lectures so far we have discussed internal energy and heat, but now we are going to formalise these definitions further and see how they relate to the first law of thermodynamics. 

The first thing we should be explicit about is what the internal energy of a system is:

- This is simply *the sum of the kinetic energy of all the constituent particles in a system plus all of the potential energies of the interactions between them*. This is the total amount of energy contained within the body (or system).

- The kinetic energy of the system moving as a whole **does not** contribute to the internal energy of the system, nor does the potential energy of the system as a whole due to external forces. 
 
- Potential energy is represented by the symbol $U$.

The internal energy of a system can be increased in one of two ways: 

1. Supplying a quantity of heat, $Q$, to the system, this will increase the internal energy by the the amount of heat supplied:

$ \Delta U = \Delta Q.$

2. Doing work on the system, for example, compressing a gas in a system by doing mechanical work upon it, this will add to the internal energy by an equivlent amount:

$ \Delta U = \Delta W $. 


In either case, the internal energy would increase by the same amount of energy as that done by doing the work or supplying the heat. This then leads us on to the formal definition of the first law of thermodynamics...

### First Law of Thermodynamics

The first law of thermodnyamics simply states that the change in internal energy of a systemis equal to the heat added to the system plus the work done *on* the system. This can be expressed mathematically as:

$ \Delta U = Q + W.$

This is simply a statement of conservation of energy i.e. energy cannot be created or destroyed, only converted from one form to another. 


```{warning}
A key feature that we must be aware of is the notation used. In our definition for the first law of thermodynamics we have said $W$ is positive. This is because we define $W$ as the work done **ON** the system, whereas other texts will define this as the work done **BY** the system, which leads to the relationship:

$ \Delta U = Q + W.$

One way to think of this is to consider gas in a piston. If the system (the gas) is doing work, for example by expanding against its surroundings, energy is leaving the system and the internal energy will decrease (assuming that no heat was supplied to the system). 

In these lectures, we will only use the expression outlined outside of this warning. When you study thermodynamics on more depth, you will see the important of this euqation in relation to processes such as engines.
```

## Heat Capacity (C)

When we consider the amount of heat added to a system to increase its temperature by a given amount, we will need to consider the heat capacity of the material in question. 

The heat capacity is simply the amount of heat energy required to raise the temperature of a body by 1K:

$ C = \frac{\Delta Q}{\Delta T}. $

Accordingly it has units of **J$\,$K$^{-1}$**. For example, if I had a block of glass, and I wanted to increase the temperature by 10 degrees, I would need to supply it with ten times its heat capacity worth of energy. 

In reality, it is actually fairly uncommon for us to use the general heat capacity when performing calculations, and this is because it is dependent on both the size, mass and materials of the system and as such must be determined for every system that we are performing calculations for. This is clearly not convinient. 

---
### Specific Heat Capacity (c) 

It us much more common for us to perform calculations using the *specific* heat capacity, which take into account the amount of material that heat is being supplied to when calculating how much energy is required to change its temperature. 

For example, the specific heat capacity is the amount of energy required, per unit mass, to increase the temperature of a body by 1K. It is therefore related to the heat capacity via:

$ c = \frac{C}{M} = \frac{\Delta Q}{m \Delta T}.$

Accordingly it has units of **J$\,$kg$^{-1}\,$K$^{-1}**$. Each material has its own specific heat capacity as shown below:

|Material|Specific Heat Capacity, c, J$\,$kg$^{-1}\,$K$^{-1}$| 
|---|---|
|Aluminium |900|
|Copper|390|
|Silver|230|
|Steel|480|
|Concrete|920|
|Brick|750|
|Polythene|2200|
|Water|4200|

Note how large the specific heat capacity of water is - this means it can store a lot of heat energy!

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

specificHeatCapacity = [230, 390, 480, 750, 900, 920, 2200, 4200]
materialDensity = [10490, 8960, 7850, 1650, 2712, 2300, 917, 998]
material = ["  Silver", "  Copper", "  Steel", "  Brick", "  Aluminium", "  Concrete", "  Polythene", "  Water (Liquid)"]

plt.figure(figsize=(10,10))
plt.plot(materialDensity, specificHeatCapacity, marker = 'o', linestyle = ' ', color = 'b')
plt.xlabel(r'Density [$\:kg \:m^{-3}\: $]')
plt.ylabel(r'Specific Heat Capacity [$\:J \: kg^{-1} \: K^{-1}\:$]')
plt.grid()

for i, txt in enumerate(material):
    plt.annotate(txt, (materialDensity[i], specificHeatCapacity[i]))

plt.show()
````

If we plot these specific heat capacities as a function of the material density we see a rough trend emerging - as the density of the material increases, the specific heat capacity of the material decreases. In a few lectures time we will look at why this is. 

Now we have an understanding of what specific heat capacity is, we can use it to perform calculations and determine how much heat energy we must supply to or remove from an object with a given mass to change its temperature by a given amount. Rearranging the specific heat capacity equation tells us how much heat energy is required to raise a mass, $m$, by a temperature, $\Delta T$, providing we know the specific heat capacity of the material under consideration:

$ c = \frac{\Delta Q}{c \Delta m} $

$ \Delta Q = m c \Delta T $.

This provides us with the link between the heat energy required, $\Delta Q$, and the temperature change, $\Delta T$, we will observe for a body of mass $m$.


```{admonition} A Note on Notation
Often you will see $d \theta$ used in place of dT. This is simply to note that the difference in temperatures has been measured in Celsius rather than Kelvin. 

This is immaterial as the two temperature scales increase in the same increments, e.g. if we increase the temperature of a body by 80˚C, we have also increased its temperature by 80 K. Therefore in our calculations it makes no matter which one we use, so long as we are looking at a change in temperature. 

Having said that, you may wish to consistently use Kelvin in your calculations as for other calculations you will perform in this unit (for example when we are working with ideal gases), the equations we use assume we are working in Kelvin. 
```

### Molar Heat Capacity (C$_{mol}$)

An alternative to the specific heat capacity is the *molar* heat capacity, which is simply the amount of energy required to increase the temperature of one mole of a substance by 1K. 

A quick reminder that "molar" simply means one mole of something, which is equal to $6.022141429 \times 10^{23}$ elementary entities (atoms or molecules) of a given substance. Therefore, one mole of a substance A will have a different mass to one mole of substance B, but they will contain the same number of elementary entities. 

The number of moles is often denotes using the letter $n$ and as such the molar heat capacity, $C_{mol}$, can be calculated as:

$ C_{mol} = \frac{\Delta Q}{n \ Delta T}.$

## Changes of State and Phase Transitions

![](Figures/WaterPhaseDiagram.png)

So far we have only considered supplying/removing energy to/from materials to change their temperatures, but we have not considered any phase changes. Lots of substances can display different states or phases and the physical properties of these states may differ despite the chemical substance remaining the same. 

The diagram above was shown in a previous lecture and illustrates the fact that phase changes are dependent on pressure and temperature in addition to the material in question, e.g. the phase diagram from water will obviously differ for oxygen. 

**Question:** How do we determine how much energy we must supply to a material for a phase change to occur? For this we must introduce a new concept, the latent heat. 

**The specific latent heat, $L$, is simply the amount of energy required to change the phase of a unit mass of material.** 

One key thing to note is that phase changes occur at a constant temperature. This is to say that if we are supplying energy to a block of ice at -10˚C, initially the temperature will increase as we supply heat, but when it gets to 0˚C and begins to melt, as we continue to supply heat the temperature of the melting ice will remain the same, then as we continue to supply energy, the resulting water will again begin to increase in temperature. 

For a given mass of a material with a latent heat of $L$, the amount of energy required to change the phase is simply equal to:

$ Q = mL_x, $

where $L_x$ would be $L_f$ for fusion and $L_v$ for vapourisation. The latent heat differs for different phase transitions, for example fusion (liquid to solid) and vaporisation (liquid to gas) have different latent heats associated with them. Some materials sublimate - sublimation is when a solid turns into a gas directly without the liquid phase between. 

As there is no temperature change associated with a change of phase, the unit for the specific latent heat is simply **J$\,$kg$^{-1}**$.

The final thing to note is that phase changes are reversible! Just as we can melt something we can freeze something! The only difference is that we are either supplying energy to or taking energy from the material, thus the internal energy of the system will either increase or decrease.

### Question 1

- a) An aluminium block, mass $0.50\,$kg, is heated by a $40\,$ W electrical heater embedded in it. In $56.3$ seconds, the temperature of the block rose from $15.0$°C to $20.0$°C. (a) Calculate the specific heat capacity of aluminium, ignoring heat losses.

- b) A kettle contains $500\,$ml of water at $20$°C. If the water is heated to $353.15\,$K in a time of 1 minute and 35 seconds. Determine how much energy the kettle must supply to the water. The specific heat capacity of water is 4200$\,$J$\,$kg$^{-1}\,$K$^{-1}$.

- c) What is the power of the kettle?

```{admonition} Q1a Solution
:class: dropdown

To answer this question we simply need to recal the definition of the specific heat capacity, which is the amount of energy it takes to increase the temperature of a substance by one kelvin per unit mass:

$c = \frac{\Delta Q}{m \Delta T},$

The change in temperature, $\Delta T$ is simply $5\,$ K, and the amount of heat energy supplied can be calculated by recalling that power is simply equal to the rate at which work is done:

$P = \frac{\Delta E}{\Delta t}, $

$\therefore \Delta E = P \Delta t,$

accordingly:

$c = \frac{P \Delta t}{m \Delta T},$

$c = \frac{40 \times 56.3}{0.5 \times 5},$

$c = 901\, J\,kg^{-1}\,K^{-1}.$

```

```{admonition} Q1b Solution
:class: dropdown

To answer this questions, we are simply going to use the equation $Q = mc \Delta T$, all we need to do is convert the units provided in the question:

- $20$˚C = $293.15\,$K,

- $500\,$ ml = $500\,$ g^{*1},

therefore, the total amount of heat energy supplied is:

$ Q = mc \Delta T, $

$ Q = 0.5 \times 4200 \times (353.15 - 293.15), $

$ Q = 126000\, J. $

^{*1} If you did not know this, 1 cubic centimetre of water weights $1\,$g and occupies $1\,$ml of volume.

```

```{admonition} Q1c Solution
:class: dropdown

To answer this question we are again going to consider the relationship between time, energy and power. We are also going to assume that the transfer of energy from the kettle to the water was a 100% efficient process i.e. no heat losses:

- 00:01:35 = $95\,$s,

$ P = \frac{\Delta E}{\Delta t}, $

$ P = \frac{126000}{95} $

$ P = 1326 \, W $
```

- c) How much energy is required to increase the temperature of the resulting water to its boiling point ($100.0$ °C)? The specific heat capacity of water is 4200$\,$J$\,$kg$^{-1}\,$K$^{-1}$.


- d) How much energy is required to convert all of the water into steam? The latent heat of vapourisation is $2258\times10^3\,$ J$\,$kg$^{-1}$.


- e) How long would it take a $1500\,$W heater to convert the initial block of ice completely into steam?

```{admonition} Q2a Solution
:class: dropdown

Again, we are just going to use the equation $Q = mc\Delta T$ to determine the energy requried to increase the temperature of the ice block from $-12$˚C to $0$˚C, which is a change in temperature of $12.00\,$K:

$ Q = mc \Delta T, $

$ Q = 1 \times 2108 \times 12, $

$ Q = 25300\, J. $
```
