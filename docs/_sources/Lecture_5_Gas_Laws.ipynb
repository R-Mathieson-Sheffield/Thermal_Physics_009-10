## Useful things to know

Before delving into the gas laws, it is important that we lay the ground rules for the units that we are going to use moving forward. The gas laws themselves concern themselves with the relationships between temperature, pressure, volume and the amount of gas, so it is important that we are using the correct units in our calculations. 

- Pressure is defined as the amount of force something exerts per unit area $\left( P = \frac{F}{A}\right)$ therefore the unit for pressure is $\text{N m}^{2}$. This is simply referred to as the Pascal which is denoted $\text{ Pa}$.

- An important number to remember is that air pressure at sea level is equal to $1.01 \times 10^5\text{ Pa}$. This is worth remembering as a reference number. 

- The gas laws also require us to measure volume in $\text{m}^3$ and temperature in Kelvin **not Celsius**. 

- When performing gas law calculations we may either use the number of molecules or moles when describing the amount of gas used in our gas law calculations, however we must make sure we use the correct gas constant for the units we have - more on this later!

## The Gas Laws

### Boyle's Law

Boyle's law states that **the pressure of a gas is inversely proportional the the volume that the gas occupies at a given temperature and number of moles**.  
This can be expressed mathematically as:

```{math}
P \propto \frac{1}{V}
```

For this law to be true, the temperature and number of moles of gas in the system must remain constant. 

What does this mean? Well, consider a piston in an engine in which we have gas contained within a cylinder, the volume of which is controlled using a piston. If we lower the piston we are reducing the volume in which the gas can occupy.  This means that the gas molecules are more likely to collide with one another and the surface of the container in which they are held, thus the pressure that they exert on the walls of the container increases. The opposite is true if we extract the piston from the chamber. In this case the volume of the chamber increases and the pressure the gas exerts on the walls of the container decreases. 

Naturally, as we have a linearly proportional relationship we can introduce a constant of proportionality, $c_1$, meaning:

```{math}
P = \frac{c_1}{V}
```

which can also be written as:

```{math}
PV = c_1
```

If the system is a closed (one meaning no gas enters or leaves the system) **and** the temperature remains constant then this constant is the same for all pressure-volume combinations meaning we can finally state:

```{math}
P_1 V_1 = P_2 V_2
```

---

### Charles' Law

Charlesâ€™ Law states **the volume that a gas occupies is directly proportional to the temperature of the gas for a given pressure and number of moles of gas**.  
Mathematically this is given by:

```{math}
V \propto T
```

For this law to be true, the pressure and number of moles of gas in the system must remain constant.

Let us again consider our gas in a piston. This time, rather than compressing the gas with the piston we are just going to supply the gas with heat so that the temperature of the gas increases. As we are again assuming that none of the gas molecules escape, as the gas heats up their kinetic energy increases and the gas expands. The only way of maintaining a constant pressure in this scenario is for the gas in the chamber to raise the piston and increase the volume of the cylinder. 

As we have another linearly proportional relationship, we can introduce a second constant of proportionality, $c_2$:

```{math}
V &= c_2 T\\
\therefore \frac{V}{T} &= c_2,
```

and consequently:
```{math}
\frac{V_1}{T_1} = \frac{V_2}{T_2}
```

A consequence of this linear relationship is that if we plot it on a graph ($V$ on the horizontal axis and $T$ on the vertical -axis), we will get a line that we can extrapolate to lower and lower temperatures. This actually implies the existence of absolute zero (a lowest possible temperature) as if we keep going to smaller and smaller volumes we will eventually get to zero, and as it is not possible to have a negative volume the corresponding temperature must be the lower limit for temperatures!

---

### Amontons' Law

Also known as the "Guy-Lussac" or "Pressure" law, Amontons' law states that **the pressure of a gas is directly proportional to the temperature of a gas for a given volume and number of moles**.  
We can express this as:

```{math}
P \propto T
```

Again consider our gas in a cylinder with a piston but this time the piston is locked in place. If we supply heat to the gas it will gain kinetic energy and the molecules will be moving faster. However, as the volume of the container is unable to expand the gas molecules are striking the walls of the container more frequently and with more force, thus the pressure that the gas exerts increases with the temperature.

As we have another linearly proportional relationship, we can introduce a third constant of proportionality, $c_3$:

```{math}
P &= c_3 T\\
\therefore \frac{P}{T} &= c_3
```

and consequently:

```{math}
\frac{P_1}{T_1} = \frac{P_2}{T_2}
```

In summary, we have three gas laws each of which assume that one other variable and the number of moles of gas in the system remains constant (i.e. no gas escapes from the system). Each law can be very convenient individually when performing calculations, but it is also more convenient to be able to combine these gas laws into one **ideal gas law**.

## The Ideal Gas Law

If we take each of the constants that we have defined above ($c_1$, $c_2$ and $c_3$) and take the product of them we get:

```{math}
c_1 \cdot c_2 \cdot c_3 = \left(PV \right) \cdot \left(\frac{V}{T} \right) \cdot \left(\frac{P}{T} \right)
```

which can be simplified to:
```{math}
:label: eq_ideal1
c_1 \cdot c_2 \cdot c_3 &= \frac{P^2 V^2}{T^2}\\
\sqrt{c_1 \cdot c_2 \cdot c_3} &= \frac{P V}{T}
```

The left hand side of this equation is made up of constants so is simply a constant itself, $R$, so we can write equation {eq}`eq_ideal1` as:

```{math}
R = \frac{P V}{T}
```

but what is $R$ equal to? To answer this we can use the fact that one mole of an ideal gas at standard temperature ($0\,^\circ\text{C}$) occupies $22.4\,$L ($0.0224\text{ m}^3$) of volume and exerts a pressure of $1.013 \times 10^5\text{ Pa}$. Plugging this into the equation above gives:

```{math}
R &= \frac{1.013 \times 10^5 \times 0.0224}{273.15}\\
    &= 8.31\text{ Pa m}^3\text{ K}^{-1}
```

The constant $R$ is known as the **molar gas constant**. This constant is defined for one mole of gas, so if we have a different number of moles, $n$, we must adust the equation accordingly:

```{math}
:label: eq_idealgasfinal
nR &= \frac{P V}{T}\\
\therefore PV &= nRT
```

This is the ideal gas equation/law! Often we will take measurements of the same system under different conditions. For a **closed system** (one wherein no gas molecules can enter or escape), $n$ and $R$ will remain constant. This means that for a closed system, we can write:
```{math}
\frac{P_1 V_1}{T_1} = \frac{P_2 V_2}{T_2}
```

---
### Alternative Ideal Gas Law

The ideal gas law is often also written as:

```{math}
:label: eq_idealgasalt1
PV = N k_B T
```

where $N$ represents the number of molecules of gas (rather than the number of moles) and $k_\text{B}$ is the Boltzman constant, which is equal to $1.38 \times 10^{23}\text{ J K}^{-1}$. This equation is entirely consistent with the one that we defined above, and we can show this by introducing Avagadro's number, $N_\text{A}$. 

We know that one mole of gas contains $\approx 6.02 \times 10^{23}$ molecules. Therefore the number of gas molecules, $N$, is simply equal to the number of moles, $n$, multiplied by $N_\text{A}$:

```{math}
N = n N_\text{A}
```

This can be substituted into equation {eq}`eq_idealgasalt1` above:

```{math}
:label: eq_idealgasalt2
PV = n N_\text{A} k_\text{B} T
```

Now if we want to get an equation that looks like the original version of the ideal gas law (equation {eq}`eq_idealgasfinal`), we need to relate the $N_\text{A} k_\text{B}$ terms with $R$. To do this we can note that both equations {eq}`eq_idealgasfinal` and {eq}`eq_idealgasalt2` have the term $PV$ on the left hand side meaning we can equate these two:
```{math}
nRT &= nN_\text{A}k_\text{B}T\\
\therefore R &= N_\text{A}k_\text{B}T
```
We can use either ideal gas law when performing calculations, it is simply a choice of what is most convinent depending on if we are given the number of moles of gas (use equation {eq}`eq_idealgasfinal`) or the number of gas molecules (use equation {eq}`eq_idealgasalt1` or {eq}`eq_idealgasalt2`). 

```{admonition} Ideal Gas Assumptions

Whenever we use the ideal gas law in a calculation, we are making a series of assumptions about the system we are modelling. The assumption being that we are working with an ideal gas in a system that has a set of specific properties:

1. The ideal gas particles are "point-like" and have no volume. A consequence of the particles are small compared to the container they are stored in and the average distance between particles.

2. The gas particles are inert and do not attract or repel one another and only collide *elastically* off one another and the walls of the container.

3. The ideal gas particles are in constant motion.

4. The walls of the container are rigid, infinitly massive and do not move.

```

In reality, no gases actually behave in this way although in some circumstances the ideal gas model is a very good approximation of how the system will behave. Looking at our ideal gas equation (equation {eq}`eq_idealgasfinal`) we would expect that the ratio of $PV$ to $nRT$, known as the compressibility factor, would remain equal to 1. It is observed that this is often the case for low gas densities, but that as the number of molecules (and therefore density) increases, the ratio diverges. This is shown in {numref}`fig_IdealGasDeviation` below.



As the density of the gas increases, there are more and more molecules occupying a given space, which increases the probability of the gases interacting with one another electrostatically. This was not accounted for in the ideal gas model. Additionally, real gases actually do have volume, so even if the dimensions of the container remain the same, as more and more gas molecules are added, there is less and less volume for them to move around in. Again, this is not accounted for in the ideal gas model. 

Finally, at high temperatures it would be expected that the kinetic energy of a real gas is high and therefore the intermolecular attraction acting between molecules is negligable in comparison - in this scenario, the real gas would behave similarly to an ideal gas. At lower temperatures, perhaps close to the condensation point, the kinetic energy of the gas particles is much lower and they move much more slowly. Now the intermolecular forces acting between molecules is no longer negligable as assumed by the ideal gas model. In this scenario, the real gas will no longer behave like an ideal gas.

## Heat, Work and Internal Energy

Recall that the internal energy of a system (e.g gas) can be increased by either doing work **on** the system, $\Delta W$, or by adding heat **to** the system $\Delta Q$, and that the internal energy of the system will increase by an equivalent amount:
```{math}
\Delta U = \Delta Q + \Delta W
```
This is the first law of thermodynamics and applies to all systems, not just gases. **The internal energy of a gas depends only on its temperature, not on its pressure or volume** (we do not cover the kinetic theory of gases on this course, but you are free to look up this proof yourself.)

A more intuitive way to think of this is to consider that the temperature of our gas is related to the kinetic energy of its molecules - this is how we defined temperature in the first lecture. We are going to consider one way of doing work on a gas, the particular case we are going to look at is **isobaric compression**. 

In thermodynamics there are lots of processes that take place where one parameter remains constant, for example temperature, volume or pressure. These are known as isothermal, isochoric and isobaric processes, respectively. 

Let us consider what happens when we compress a gas. In order to do this we must do work on the gas. If we change the volume of the container the gas is contained within by a small amount $\Delta x$, the pressure will remain approximately constant. If we assume that the piston compressing the gas moves freely (it moves without being impeded by friction), we may use our classic equation for work done to determine the work done on the gas as it is compressed:

```{math}
:label: eq_WorkDoneGas1
W = F \Delta x
```

When considering gases we are concerned with the pressure and volume. The pressure is defined as
```{math}
:label: eq_WorkDoneGas2
P &= \frac{F}{A}\\
\therefore F &= PA
```
````{margin}
```{note}
You may ask why we do not consider the whole surface area of the container as the gas is interacting with it all. Here we are considering the work done by the piston so only need to consider this area.
```
````
where $A$ is the cross sectional area of the piston. The change in volume associated by compressing the piston by amount $\Delta x$ is
```{math}
:label: eq_WorkDoneGas3
\Delta V &= A \Delta x\\
\therefore \Delta x &= \frac{\Delta V}{A}
```

Substituting equations {eq}`eq_WorkDoneGas2` and {eq}`eq_WorkDoneGas3` into equation {eq}`eq_WorkDoneGas1` gives expression for the work done in compressing the gas in terms of pressure and volume:

```{math}
W &= PA \cdot \frac{\Delta V}{A}\\
    &= P\Delta V
```

This means we can determine how much work must be done on a gas to compress it by some volume change (assuming that the pressure remains constant). Next year, you will see more complex calculations to determine the work done when the pressure is not longer constant.

## Molar Heat Capacities

Finally, now we will introduce the idea of heat capacities at constant pressure and at constant volume. Recall that the molar heat capacity is the amount of heat required to raise one mole of a substance by one Kelvin. However, in different circumstances, the amount of energy required to increase the temperature of the gas will differ. 

Again consider a gas contained within an engine cylinder. If we supply heat to a gas which is held at a constant volume (the change in volume is zero because the piston is unable to move), the gas will do no work on its surroundings (as there is no change in volume - see equation above). This means that all of the heat added solely goes into increasing the temperature of the gas. 

On the other hand, if we are to supply heat to a gas and allow it to maintain a constant pressure, this means that the gas must expand and occupy a greater volume. Using the cylinder example above, this means that the gas is going to push against the surrounding walls and move the piston (which can freely move in this scenario). In this instance the heat supplied both increases the temperature of the gas, which then does work on the surroundings. This means that we are going to need to supply more heat for the same increase in temperature when compared to the constant volume scenario. 

An important relationship to remember is that the molar heat capacity at constant pressure, $c_p$, is equal to the molar heat capacity at constant volume $c_v$ plus the molar gas constant:

```{math}
c_p = c_v + R
```

``````{admonition} Question 5.1 - Plastic bottle under pressure
:class: dropdown, hint

`````{tab-set}
````{tab-item} Question
An empty $500\text{ ml}$ bottle is opened at the top of a mountain, it is subsequently sealed and brought back down the mountain. As the air external air pressure increases, to maintain a pressure equilibrium inside and outside of the bottle, the bottle becomes compressed to a new volume of $256\text{ ml}$. Assume that the temperature of the gas remained constant at $25\,^\circ\text{C}$.

- a) Determine the ratio of the pressure at the top of the mountain to the bottom of the mountain.
- b) If the initial gas pressure within the bottle at the top of the mountain was $7.60\text{ kPa}$, determine the air pressure at the bottom of the mountain.
- c) Calculate how many gas molecules are contained within the sealed bottle.
````

````{tab-item} Hint
**Part a**
* What happens to the volume of the bottle and the external pressure acting on it as it descends the mountain?
* What gas law describes the relationship between pressure and volume when the temperature is constant?

**Part b**
* You know the pressure at the top of the mountain and the ratio of pressures. How can you use this information to find the pressure at the bottom of the mountain?

**Part c**
* Which gas law relates pressure, volume, temperature, and the number of gas molecules?
* You have the pressure, volume, and temperature at the bottom of the mountain. 
* Can you use this information to calculate the number of gas molecules? 
````

````{tab-item} Solution
**Part a**  
For this question we can assume that the number of moles of gas trapped in the bottle remains constant and the gas itself behaves like an ideal gas. We also know that the temperature of the system remains constant at $25\,^\circ\text{C}$ which is $293.15\text{ K}$. Therefore we can use Boyles' law relating pressure and volume at the top ($T$) and bottom ($B$) of the mountain:

```{math}
P_T V_T &= P_B V_B\\
\therefore \frac{P_T}{P_B} &= \frac{V_B}{V_T}\\
    &= \frac{256}{500}\\
    &= 0.512
```


**Part b**  
As we have now determined the ratio of the pressure at the top of the mountain to the pressure at the bottom and we already know the pressure at the top, we can nicely determine the pressure at the bottom of the mountain:

```{math}
\frac{P_T}{P_B} &= 0.512\\
\therefore P_B &= \frac{P_T}{0.512}\\ 
    &= \frac{7.60 \times 10^3}{0.512}\\
    &= 14800\text{ Pa} = 14.8\text{ kPa}
```

**Part c**
For this question, we know the temperature of the gas, and we have two sets of pressure and volume measurements we can use to determine the number of gas molecules using the ideal gas equation in the form of number of molecules $N$ rather than number of moles $n$ as we are trying to calculate the former. In an exam I would be tempted to use the pressure and volume provided rather than calculated just incase I had made a mistake with my earlier calculations:

```{math}
PV &= Nk_BT\\
\therefore N &= \frac{PV}{k_BT}\\
    &= \frac{17.6 \times 10^3 \times 0.0005}{1.38 \times 10^{23} \times 293.15}\\
    &= 9.24 \times 10^{20} \text{ molecules}.
```
The trick with this question is being able to convert between ml and cubic metres $(1\text{ ml} = 1\times10^{-6}\text{ m}^3)$.
``````
