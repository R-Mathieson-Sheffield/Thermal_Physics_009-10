## Heat Transfer Mechanisms

There are three mechanisms by which heat may be tranferred: **conduction, convection** and **radiation**. The mechanism of heat transfer is broadly dependent on the state of the material. In solids, we tend to find that conduction will dominate, whereas in liquids and gases, the primary mechanism is via convection.

## Conduction

Let us start with a rather fundamental question about how heat conducts through a material. **What is actually happening at an atomic level when heat is being conducted?**

Metals are examples of good thermal conductors. If we place one end of a metal rod in a flame, the temperature of that end will increase, the temperature of the rest of the rod will also start to increase - why is this?  
The atoms in the hot end of the rod are at a higher temperature and therefore have a higher kinetic energy. This means that the magnitude of the oscillations of the atomic vibrations is increased compared to the cooler atoms. 

The hotter atoms then jostle with the cooler atoms and transfer to them some of their kinetic energy. This process continues along the material allowing the heat energy to be transferred along the bar. 
````{margin}
```{note}
This wiggling about a fixed point is somewhat akin to a pendulum. A lower temperature atom is like a pendulum with a low amplitude whereas the higher temperature atoms are like a larger amplitude pendulum. The pendulum system itself doesn't travel along some axis (assuming it's not on wheels or haunted), even though the pendulum bob is moving back and forth.
```
````
It is key to note that the atoms in this scenario do not move **along** the bar, but the energy does. Instead the atoms are wiggling with a greater amplitude about some fixed point.

One of the reasons that metals are such good conductors is that they are formed of an atomic lattice with many delocalised electrons. Unlike the atoms, these electrons are able to more freely move throughout the bulk of the material, which means they are readily able to carry energy from the hotter regions to the cooler regions.

### Coefficient of Thermal Conductivity

We know that some materials conduct heat more readily than others, but why is this and what other factors contribute? When we are considering how well a material conducts heat, we are effectively trying to determine the rate at which **heat flows** through the material. As you may have noticed by now when we are looking at relationships between variables for different materials, we often refer to a property that defines the relationship (often a constant of proportionality) with values that are unique to the material in question. In this instance we are going to consider the **coefficient of thermal conductivity**. 
Let us derive the relationship for heat flow that lets us determine the coefficient of thermal conductivity for any given material. 

To begin we should consider the factors that would impact heat flow through a material. To simplify our thought experiment we are consider a circular disk, with an area $A$, and a thickness $\Delta x$, with a temperature gradient $\Delta\theta$ across the thickness. The temperature gradient between the two circular surfaces of the disk is key as heat must flow from hot to cold. 

```{figure} Figures/CoefficientConductivity.png
:name: CoefficientConductivity
:alt: A schematic of one-dimensional heat conduction through a disk. The disk is shown with a temperature gradient across it. The left side of the disk is labeled with a higher temperature (θ + Δθ) and the right side with a lower temperature (θ). An arrow passes through the disk from left to right, representing the flow of heat (ΔQ/Δt). The thickness of the disk is labeled as Δx and its surface area is labeled as A.
:width: 80%
:align: center
A schematic representation of one-dimensional heat conduction through a disk of surface area A and thickness $\Delta x$. The temperature gradient across the disk is represented by $\theta + \Delta\theta$ and $\theta$. The heat flow rate $\frac{\Delta Q}{\Delta t}$
is shown as an arrow passing through the disk.
```

So we want to consider how these factors affect the flow of heat $\Delta Q$ in some time interval $\Delta t$, i.e. $\frac{\Delta Q}{\Delta t}$:

| Property | Expectation | Model |
|---|---|:-:|
|**Area** | If we have a larger area for heat to flow through, the rate of heat flow will be increased. | $\frac{\Delta Q}{\Delta t} \propto A$ |
|**Material thickness**| If the material is thinner, the rate of heat flow will be increased. | $\frac{\Delta Q}{\Delta t} \propto \frac{1}{\Delta x}$ |
|**Temperature Gradient**| If there is a larger temperature difference, $\Delta \theta$ between the ends of the disk, the rate of heat flow will be increased for a given thickness. | $\frac{\Delta Q}{\Delta t} \propto \Delta\theta$ |




So with this in mind, how would we build an expression for the rate of heat flow, $\frac{\Delta Q}{\Delta t}$, that takes into account each of these contributing factors? Assuming the variables involved (area, thickness and temperature gradient) are independent then we can multiply all of the models in the above table into a single model

```{math}

\frac{\Delta Q}{\Delta t} \propto A \times \frac{1}{\Delta x} \times \Delta \theta\\
    &\propto A\frac{\Delta \theta}{\Delta x}
```

What do we do when we have a proportional relationship? We introduce a constant of proportionality! Therefore:

```{math}
:label: eq_TPS6_heatflow
\frac{\Delta Q}{\Delta t} = -\lambda A\frac{\Delta \theta}{\Delta x}
```

where $\lambda$ is the coefficient of thermal conductivity which has units of $\text{W m}^{-1}\text{ K}^{-1}$. **The negative sign comes from the fact that the temperature gradient is negative, but the heat flow must be positive.** The coefficients of conductivity for various materials are tabulated below:

|Substance|$\lambda$ W$\,$m$^{-1}\,$K$^{-1}$|
|---|---|
|Copper|$390$|
|Aluminium|$220$|
|Ice (@ -4˚C)| $2.1$|
|Window Glass|$0.93$|
|Water|$0.60$|
|Air|$0.024$|

We can see here that there is trend wherein the density of the material is broadly proportional to the coefficient of conductivity. This makes sense when we consider the microscopic process of conduction discussed above. We know that metals are particularly good conductors of heat due to the fact that they are solids and allow molecular vibrations to effectively transmit energy along the material and they have an excess of dissociated conduction electrons, which is the same reason that they are such effective conductors of electricity. 
Temperature also affect the ability of a material to conduct heat. Some materials are good conductors at low temperatures, but as they increase in temperature their physical properties change which means they do not as effectively conduct heat. 

Just to reinforce this concept, here we can compare the thermal conductivities of two different materials, copper and fused quartz. We can see that copper is an excellent conductor of electricity and heat, whereas the same is not true for fused quartz. This makes sense as copper has many delocalised electrons, whereas quartz does not. 

|Material|Thermal Conductivity W$\,$m$^{-1}\,$K$^{-1}$| Electrical Conductivity ($\Omega^{-1}\,m^{-1}$)|
|---|---|---|
|Copper|$390$|$5.9 \times 10^{7}$|
|Fused Quartz|~$2$|~$1 \times 10^{-17}$|

### Temperature Distribution Along a Bar

Now that we understand the mechanics of conduction, I would like to briefly make you aware of how the temperature across a metal bar varies during conduction. 

To do this I first want you to consider this scenario here. We have a metal bar connected to a **heat source** and a **heat sink**. A heat sink is just something that draws heat energy out of a system to help keep it cool. 

```{figure} Figures/MetalBar.png
:name: fig_MetalBar
:alt: Four stages showing the flow of heat through a metal bar. In the first stage, the left side of the bar, labeled "Heat Source," is red, indicating a high temperature. The right side, labeled "Heat Sink," is blue, indicating a low temperature. The rest of the bar is a neutral color. In the second stage, the red area has expanded slightly towards the right, showing that heat has begun to flow through the bar. In the third stage, the red area has expanded further to the right, indicating that more heat has been transferred. In the fourth stage, the red area has expanded even further towards the right, showing that heat continues to flow from the heat source to the heat sink.
:width: 80%
:align: center
A series of four stages illustrating heat conduction through a metal bar over time. Heat is applied to one end of the bar (heat source) and flows towards the other end (heat sink). The colour gradient within the bar represents the temperature distribution, with warmer colours indicating higher temperatures.
```


- **Step 1:** Initially the heat source and the heat sink are the same temperature.  
- **Step 2:** The heat source is activated and begins to supply heat to the bar. Initially one end of the bar will heat up and heat will slowly be transferred along the length of the bar. 
- **Step 3:** If we were to place temperature probes along the length of the bar we would see that the temperature at each point varies as a function of time. 
- **Step 4:** Eventually a **steady state** is reached and we obtain our final temperature distribution. The way this distribution looks will be dependent on whether or not the metal bar is **unlagged** or **perfectly lagged**.

```{note}
A temperature distribution is simply a snapshot of what the various temperatures are along the length of the bar. We can plot this with temperature on the vertical axis and the position along the bar on the position axis. 
```

Lagged and unlagged heat flow may be represented using **heat flow lines**, which are simply a useful visualisation of how heat energy is transferred through a medium. The line density is proportional to the heat flow rate and the direction of the line indicates the direction that heat is flowing.

```{figure} Figures/LaggedUnlagged.png
:name: fig_LaggedUnlagged
:alt: Two diagrams illustrating heat flow through a bar. The left diagram represents perfectly lagged heat flow. Two red rectangles represent the bar. Red arrows show heat flowing in straight lines from the Heat Source on the left side of the left rectangle to the Heat Sink on the right side. Below the diagram, a line graph shows a linear decrease in temperature from left to right. The right diagram represents unlagged heat flow. Two red rectangles represent the bar. Red arrows show heat flowing from the Heat Source on the left side of the right rectangle towards the Heat Sink on the right side. Some arrows curve outwards, indicating heat loss to the surroundings. Below the diagram, a line graph shows a curved decrease in temperature, with a steeper drop at the beginning and a more gradual drop towards the end.
:width: 100%
:align: center
A diagram comparing perfectly lagged and unlagged heat flow through a bar. The left side shows perfectly lagged heat flow, where heat flows in a straight line from the heat source to the heat sink, resulting in a linear temperature gradient. The right side shows unlagged heat flow, where heat is lost to the surroundings, resulting in a curved temperature gradient.
```


We can see that in the case of unlagged heat flow, the heat flow lines are parallel and all of the heat is transferred from one end of the bar to the other. In this case if we plot the temperature measured by temperature probes across the bar as a function of distance, we will see that there is a linear relationship (i.e. a constant temperature gradient).


This is not the case for the lagged heat flow where the heat flow lines are not parallel and are instead directed towards the sides of the bar. This means that the bar will radiate heat energy away from the sides of the bar. Again if we were to plot the temperature as a function of distance along the bar, we would see that the temperature gradient across the bar is not constant. This makes sense, as the further along the bar you take a measurement, the more surface area the heat has been able to radiate away through.

Now that we understand the difference between lagged and unlagged heating we can consider how we may experimentally determine the coefficient of thermal conductivity of a material using **Searle’s Bar Method**.

### Searle's Bar Method

The Searle's bar method is one for determining the coefficient of thermal conductivity of a material. The method assumes that the bar is lagged so that the temperature gradient across the bar is constant. The setup for the experiment is as follows:

```{figure} Figures/SearlesBar.png
:name: fig_SearlesBar
:alt: A diagram of Searle's apparatus. The apparatus consists of a long, rectangular metal bar. Steam flows through a jacket surrounding the left end of the bar, heating it up. Thermocouples are attached at various points along the length of the bar to measure the temperature. A jacket surrounding the right end of the bar allows cold water to flow through, cooling that end. The diagram also shows labels for the steam inlet and outlet, warm water outlet, cold water inlet, and the distances between the thermocouples.
:width: 80%
:align: center
A schematic diagram of Searle's apparatus for measuring the thermal conductivity of a metal bar. The apparatus consists of a metal bar surrounded by a steam jacket to heat one end. Thermocouples are attached along the length of the bar to measure the temperature gradient. Cold water flows through a jacket surrounding the bar to remove heat from the other end.
```

We have a metal bar of unknown thermal conductivity connected to a heat supply, in this instance a continuous flow of steam passing over one end of the bar to heat it, and a heat sink at the opposite end of the bar, in this case a continuous supply of cool water. 

We place temperature probes on the bar at two points separated by a known distance, and also place temperature probes in positions that measure the temperature of a known mass of water before it has been and after it has been heated by the bar. 

Mathematically, we can define the linear temperature gradient across the bar simply as the change in the temperature for a given length:

```{math}
\frac{\Delta \theta}{\Delta x} = \frac{\theta_2 - \theta_1}{l}
```

where $l$ is the distance between the two temperature probes. **Remember that the temperature of the bar is going from hot to cold, so as you move along the bar the temperature will become cooler. Therefore the temperature gradient should be negative!**

Another parameter we can measure from this experiment is the rate at which heat is being transferred to the water that is cooling the far end of the bar. If the temperature of the water changes by $\Delta \theta$, then the amount of heat required to increase the temperature of the water is simply given by:

```{math}
Q = mc \Delta \theta
```

where $c$ is the [specific heat capacity](TPS_Topic4_SpecificHeatCapacity) of water. The rate of heat transfer is obtained by dividing both sides of this equation by the change in time over which the change in temperature took place: 

```{math}
\frac{\Delta Q}{\Delta t} = mc \frac{\Delta \theta}{\Delta t}
```

which can be expressed as:

```{math}
\frac{\Delta Q}{\Delta t} = mc \frac{\theta_4 - \theta_3}{\Delta t}
```

At this point we now have two very familiar expressions, one related to the heat flow and one related to temperature gradient - how might we use these to determine the coefficient of thermal conductivity of the bar? We have already studied the heat flow through a material which led to equation {eq}`eq_TPS6_heatflow`. Here the thickness of the material is the distance separating the two temperature probes on the bar, i.e. $\Delta x = l$ and therefore

```{math}
mc \frac{\theta_4 - \theta_3}{\Delta t} = -\lambda A\frac{\theta_2 - \theta_1}{l}
```

which can then be rearranged to make $\lambda$ the subject:

```{math}
\lambda = - \frac{mcl}{A\Delta t}\frac{\left(\theta_4 - \theta_3 \right)}{\left(\theta_2 - \theta_1 \right)}
```

```{warning}
You are expected to know and understand this method for determining the coefficient of thermal conductivity of a material (including the derivation). It would be worth working through the steps of the derivation above yourself to ensure that it makes sense to you.
```

### Thermal Conductivity: Combined Materials

The final thing that we are going to consider in this lecture is how we calculate the thermal conductivity of composite materials, e.g. one layer of material placed on top of another as shown below:

```{figure} Figures/CompositeMaterial.png
:name: fig_CompositeMaterial
:alt: A diagram showing a rectangular block composed of two layers. The top layer is pink and the bottom layer is blue. The top layer is labeled with a subscript 1 and the bottom layer is labeled with a subscript 2. The top and bottom layers have different thicknesses, labeled $\Delta x_1$ and $\Delta x_2$, respectively. Each layer is also labeled with a symbol $\lambda$, representing its thermal conductivity. An arrow pointing upwards through the center of the block represents the flow of heat (Q) from the bottom to the top. The temperature at the bottom of the block is labeled $\theta_1$ and the temperature at the top of the block is labeled $\theta_2$.
:width: 80%
:align: center
A schematic diagram illustrating steady-state heat flow through two materials of different thermal conductivities ($\lambda_1$ and $\lambda_2$) and thicknesses ($\Delta x_1$ and $\Delta x_2$). The temperature at the bottom of layer 2 is represented by $\theta_1$ and the temperature at the other end is represented by $\theta_2$.
```

There is a simple formula to model this, but in order to do so we need two things. Firstly we are going to make a direct comparison to resistors in circuits as an analogy. The total resistance is equal to the sum of the two, as the current needs to flow through one and then the other. This is equivalent to the system we have here but instead of resistance to electical flow we are dealing with resistance to heat flow.

This leads us onto the second thing we need. So far we have considered the thermal **conductivity** which quantifies how well heat flows in a material, whereas the resistance in a circuit quantifies the **resistance** to current flow (as opposed to how conductive the wire / component is).  
We can define the **thermal resistance** $R_\text{thermal}$ as
```{math}
R_\text{thermal} = \frac{\Delta x}{A\lambda}
```
which is equivalent to the electrical resistance $R_\text{electrical} = \frac{\Delta x}{A\rho}$.

We know that the total resistance of a two-component system in series is given by
```{math}
R_T = R_1 + R_2
```
which means the total thermal resistance is
```{math}
\frac{\Delta x_T}{A_T\lambda_T} = \frac{\Delta x_1}{A_1\lambda_1} + \frac{\Delta x_2}{A_2\lambda_2}
```
If both components have the same area then $A_T = A_1 = A_2$ which simplifies the above to
```{math}
\frac{\Delta x_T}{\lambda_T} &= \frac{\Delta x_1}{\lambda_1} + \frac{\Delta x_2}{\lambda_2}\\
\therefore \frac{\lambda_T}{\Delta x_T} &= \frac{\lambda_1\lambda_2}{\Delta x_1\lambda_2+\Delta x_2\lambda_1}
```

Now we can modify equation {eq}`eq_TPS6_heatflow` slightly to include $\lambda_T$ in place of $\lambda$ and $\Delta x_T$ instead of $\Delta x$, and use the rearranged expression above for  $\frac{\lambda_T}{\Delta x_T}$ which gives:
```{math}
\frac{\Delta Q}{\Delta t} &= -\frac{\lambda_T}{\Delta x_T} A \Delta \theta\\
        &= - \left(\frac{\lambda_1\lambda_2}{\Delta x_1\lambda_2+\Delta x_2\lambda_1}\right) A \Delta \theta
```

`````{admonition} Question 6.1 - Heat flow through a window
```{tab-item} Solution
This question simply requires us to rearrange the heat flow equation and solve for $\Delta x$. In the question text we are given the rate of heat flow $2723\text{ W}$, and a Watt is Joules per second so this would be equivalent to $\Delta Q=2723\text{ J}$ when $\Delta t=1\text{ s}$.

\begin{align*}
\frac{\Delta Q}{\Delta t} &= -\lambda A\frac{\Delta \theta}{\Delta x}\\
\therefore \Delta x &= -\lambda A \Delta \theta \cdot \frac{\Delta t}{\Delta Q}\\
    &= -\lambda A \left( \theta_f - \theta_i\right) \cdot \frac{\Delta t}{\Delta Q}\\
    &= -0.75 (2.2 \times 0.9) \cdot (2 - 24) \cdot \frac{1}{2723}\\
\therefore \Delta x &= 0.012\text{ m}
\end{align*}

Note: We did not need to convert from $\,^\circ\text{C}$ to Kelvin for this question as we are working with a temperature **difference**, which would be the same regardless of which temperature scale was being used. However if the temperatures were quoted in Fahrenheit you'd need to either convert or use different values for the thermal conductivity (in $\text{ W m}^{-1}\,^\circ\text{F}^{-1}$).
`````
