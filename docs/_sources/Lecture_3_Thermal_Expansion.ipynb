## Thermal Expansion

In the last lecture we began to look at thermometry which is based on measuring thermetric properties. One such thermetric property is that of thermal expansion and contraction - the change in dimensions of an object as a function of its temperature. This is one of the topics that we are going to look at today. 

Understanding thermal expansions is not only limited to thermometry, but also in everyday life and how we engineer the world around us. For example, consider the climate in the UK, the coldest temperature ever recorded was -27.2˚C and the hottest ever recorded temperature was 40.3˚C. This is a temperature difference of 67.5˚C!

When engineers are designing infrastructure projects, considerations must be taken to account for the fact that the infrastructure will need to operate in vastly different temperature ranges without failure.  An example of this is shown below where the design of a bridge incorporates an expansion joint, which prevents the bridge from buckling in high temperatures as it has been designed to accommodate a thermal expansion. 

![](Figures/ThermalExpansionExamples.png)

Similarly, we can see an example of where train tracks have buckled. Network rail aims to prevent buckling by painting some rails paler colours so they absorb less heat, reduce the speed at which trains can travel over rails in hot conditions and introduce small gaps in the rails for them to expand into. Even so this is not always sufficient, and tracks may need replacing, which means that trains are unable to run and the efficiency of the network is reduced.

Let us now consider the physics that is actually taking place when materials undergo thermal expansion. 

To understand thermal expansion, we must look at the bulk materials on a much smaller scale. Let us consider a block of metal - if we zoom in on the structure of the metal we will find atoms arranged within a lattice (a repeating structural unit of atoms):

![](Figures/CuLattice.png)

As these atoms have energy, they will vibrate around their equilibrium positions. If these are supplied with more heat (e.g. from the environment on a hot sunny day), the temperature of the metal will increase. 

As we learnt in the last lecture, temperature is directly related to the kinetic energy of the atoms, and if this increases, so does the amplitude of the atomic oscillations. This results in a larger average distance between the atoms and correspondingly and expansion of the metal. 

We can again assume that over a given temperature range, the amount of expansion is linearly proportional to the temperature of the material. We can also assume that the expansion would be proportional to the original length of the material. These assumptions can be expressed mathematically as:

$ \Delta L \propto L_0 \Delta T, $

where $\Delta L$ is the change in length, $L_0$ is the intial length and $\Delta T$ is the change in temperature. When we see a proportionality sign, we should introduce a constant of linear proportionality, which in this case is the constant of thermal exapansion, $\alpha$:

$ \Delta L = \alpha L_0 \Delta T, $

or 

$ L_f - L_0 = \alpha L_0 \Delta T, $

$ L_f = \alpha L_0 \Delta T + L_0, $

$ L_f = L_0 \left(1 + \alpha \Delta T\right) $

**We must be aware that as was the case in our thermometer example, this condition is only true over a given range of temperatures. For much larger changes in temperature, we may no longer be able to assume linear proportionality.**

---
### Coefficient of Thermal Expansion

The coefficient of thermal expansion of a given material can easily be calculated by rearranging the equation above:

$ \alpha = \frac{\Delta L}{L_0 \Delta T} $

If we rearrange our equation we can see that $\alpha$ is simply a measure of the fractional change in length for a given change in temperature. Again, we are making an assumption that this coefficient of thermal expansion is constant (whereas in reality it is dependent on the temperature range, the initial temperature and other external conditions such as pressure). We will neglect this in our calculations unless otherwise stated.

We can also see from the equation above that $\alpha$ has units of inverse temperature or "per Kelvin" (K$^{-1}$). Examples of the thermal expansion coefficients for various materials are provided below:

|Material|$\alpha$ (K$^{-1}$) @ 20˚C|
|---|---|
|Aluminium|$23.1 \times 10^{-6}$|
|Brass|$19 \times 10^{-6}$|
|Concrete|$1 2\times 10^{-6}$|
|Glass|$8.5 \times 10^{-6}$|
|Lead|$29 \times 10^{-6}$|
|Mercury|$61\times 10^{-6}$|
|Titanium|$8.6 \times 10^{-6}$|

It is worth noting that this equation holds for **all dimensions** of the material during expansion, (e.g. it could be the width, the length, the height) so long as this is a linearly expanding material. This may not always be the case, for example, some crystals when heated will expand by different amounts in different directions. We are not going to consider this additional complication.

It is often necessary to use materials with similar values of thermal expansivity to prevent any thermal expansion from causing undue stresses on those materials (particularly for structural/machine components that are thermally cycled). 

```{admonition} Important

The equation to determine the expansion of an object for a given change in temperature is: 

$ \Delta L = \alpha $L_0$ \Delta T, $

where $\alpha$ is the coefficient of thermal expansion and has units of K$^{-1}$.
```

---
## Revisiting Fixed Points

Before we move onto to looking at the different types of thermometer, it is worth revisiting the fixed points that we have used to define our tempeature scales thus far. In our previous example of making a liquid in glass thermometer using water, we stated that the two fixed points to be used were the freezing and boiling points of water, which we assigned to 0˚C and 100˚C respectively. Looking at the diagram, can you tell me in which circumstances this is actually the case? 

![](Figures/WaterPhaseDiagram.png)

This plot shows us the pressure on the y-axis and the temperature on the x-axis, and the different colours represent the different phases of water. The boundaries represented by the black lines represent the temperatures and pressures that phase changes occur at under different circumstances. 

We can see that the boiling and freezing points of water actually only occur at 0˚C and 100˚C at 1 atmosphere of pressure, and that if we increase the pressure the temperatures at which these phase transitions occur change (this is most easily seen for the boiling point).

There are two other interesting points on this plot, one is the *triple point of water*, which is the temperature and pressure at which all three states of matter coexist. 

The second is the *critical point*, which is the point at which the differences in the physical properties (such as the density or the compressibility) between the liquid and vapour phase of the water become zero and the distinction between the liquid and vapour phase disappears. 

Taking this into consideration, it means that when we are defining the temperatures of our fixed points, there are other variables that we must be able to control for them to be truly accurate. Wouldn’t it be better if there was a more absolute temperature scale that we could use instead? Introducing the Kelvin scale!

---
### The Kelvin Scale

For the lower point we are going to recall that heat is simply equal to energy, and this energy could be extracted from an object continually to the point where there is no more energy left in the object to remove. At this point, the object will have a minimal internal energy. **We define the temperature of the object to be absolute zero - it is not possible to cool an object below this point**. 

It is often said that at absolute zero all molecular motion ceases, however due to the quantum nature of the universe at this scale, this is not actually true... 

Another quirk of absolute zero is it is actually impossible to cool an object to absolute zero in a finite number of thermodynamic steps. For example, consider if you already had an object at absolute zero and brought into contact with a body you wanted to cool to absolute zero, the two bodies would heat and cool until they reach thermal equilibrium, which will be above absolute zero. Whislt it is not possible to cool anything down to absolute zero, we have been able to cool to 38 pK (38 $\times 10^{-12}$ K!

**The lower fixed point for the Kelvin scale is absolute zero!**

**The upper fixed point of the Kelvin scale is the triple point of water (0.01˚C, 273.16 K at 611.657 Pa).** 

Not only is the triple point of water occur at the same temperature, but this also means that we get a very simple relationship between the celsius scale and Kelvin scale. An increase in temperature of 1˚C is equal to an increase in 1K. If we need to convert between Kelvin and Celcius, we simply add 273.15 to our celsius value. 

For example, 0˚C is simply equal to 273.15K, 20˚C is simply equal to 293.15K. Note that we do not say "degrees Kelvin", simply Kelvin!


---
## Types of Thermometer

Let us consider the different types of thermometer, with their advantages and disadvantages. The choice of thermometer you would use is dependent on many circumstances, for example, how hot are we measuring something to be? Is the temperature probe going to be incorporated into a small device? etc.

When we defined our temperature scale, we assumed that the thermometric property we were measuring was changing linearly with the temperature. On this basis, we used two fixed points to define an equation that gave us a relationship between the temperature and thermometric property, however **most thermometric properties don’t change linearly with temperature, or they only do so over a limited range!**

---
### Liquid in Glass Thermometer

We know that these operate via the expansion of a liquid as a function of temperature, the liquids most commonly being used are mercury or ethanol, due to the fact that both of these liquids will undergo phase changes, they are only able to measure over a limited range of temperature. 

The design of these thermometers means that they are very sensitive to changes in temperature. Simply consider a cylinder with a fixed cross-sectional area containing a liquid. If the cylinder has large cross sectional area, as the volume of the liquid expands the increase in the length of the liquid in the cylinder will barely change, whereas if the cross sectional area of the tube is very narrow, a small fluctuation in temperature will correspond to a large change in the length of the liquid! This means that these thermometers are very sensitive to changes in temperature. The change in length of the material can be calculated as:

$\Delta L = \frac{\Delta V}{A}. $

**Advantages:**
- They are portable.
- They allow for direct reading.
- There is no need to re-calibrate over the lifetime of the device.
- They are relatively cheap.

**Disadvantages:**
- They are not very accurate.
- They only measure temperatures over a limited range.
- Mercury is toxic if the thermometer breaks.

---
### Platinum Resistance Thermometer

As we have discussed previously, the electrical resistivity of a material is a thermometric property and the one that is exploited in platinum resistance thermometers - this is to say that the measured resistance of the device changes as a function of its temperature. There are two common standards:

- PT100 has a resistance of 100 Ω at 100°C.
- PT1000 has a resistance of 1000 Ω at 100°C.

![](Figures/PTResistance.png)

Platinum is chosen because of its relatively high resistivity and wide temperature range as well as for being stable, this means that the temperature probe is able to operate over a large range of temperatures with a relatively high sensitivity to changes in temperature and accuracy when measuring temperatures. The final advantage is that it is a digital device which means that it may be more useful for recording temperatures in an experiment.

The conversion between the resistance and temperature for a PT100 thermometer is: 

$ R = (0.385 \Omega ˚C^{-1}) T + 100 $

The conversion follows a simple $y = mx + c$, meaning that the response is linear. 

**Advantages:**
- They can operates over a wide range of temperatures.
- They are very sensitive and accurate.
- They allow for direct temperature to voltage conversion (digital means no need to take manual measurements).


**Disadvantages:**
- The coil type has a very slow response.
- At low temperatures, resistive heating can lead to inaccuracies.

---
### Thermistor Resistance Thermometer

A similar resistive thermometer is a thermistor, however unlike the previous resistive thermometer the temperature response is not linear. 

![](Figures/ThermistorResistance.png)

**Advantages:**
- They can have a very small size.
- They have a very quick response time.
- They allow for direct temperature to voltage conversion

**Disadvantages:**
- They need to be calibrated regularly.  
- Their temperature response is not linear.

---
### Thermocouple Thermometer

The thermocouple thermometer is another type of digital thermometer that is formed of two pieces of electrically conductive metal which have different electrical properties. These metals form a junction. As a result of the Seebeck effect, these devices produce a temperature dependant voltage at the junction.

The Seebeck effect is simply an effect wherein an EMF forms across two points of a conductor when they are at different temperatures. The magnitude of the EMF is temperature dependent which means that this thermometric property can be used for thermometry in a thermocouple thermometer.  The voltage, V, is used to calculate a distant temperature, provided that a reference temperature is known.

![](Figures/Thermocouple.png)

**Advantages:**
- They have a small size.
- They have a rapid response.
- They can operate over a very wide range of temperatures.
- They are useful in instrumenting “hot” objects.


**Disadvantages:**
- They are not as accurate as platinum resistance thermometers.
- Measured temperature determined using a reference temperature that itself is often determined using another thermometer. Therefore the accuracy of the measured temperature is dependent on the accuracy of the measured reference temperature.

---
### Constant Volume Gas Thermometer

In order to understand how a constant volume gas thermometer works, we must first understand one of the gas laws, which we will visit later in this course. Effectively Gay-Lussac’s Law states that the pressure of a gas held within a constant volume will increase as a function of temperature and that the pressure is directly proportional to the Kelvin temperature. 

![](Figures/ConstantGasVolume.png)

Using this law we are able to make measurements of temperature using the setup shown. Here we have a bulb containing a gas submerged in the environment we are going to measure. This gas bulb is directly connected to a column of mercury that is again directly connected to a second reservoir of mercury. 

We aim to maintain the level of the mercury in the first column (A) relative to an external scale by moving the mercury in the second column (B). As the gas heats, it pushes the mercury in the first column down and increases the level of the mercury in the second column. Once equilibrium has been reached, the second column (B) is physically raised until the mercury in the first column returns to its original position. This ensures that as the gas heats up, the volume it occupies remains constant and the pressure of the gas increases.  The difference in height of the mercury of the second column (B) is then recorded. 

This means that the height of the mercury in the second reservoir is directly proportional to the pressure of the gas in the fixed volume, which itself is directly proportional to the temperature of the environment being measured. 

I recommend watching [this clip](https://www.youtube.com/watch?v=n0xAQXL905c) of somebody using one of these thermometers to help clarify the explaination above.

**Advantages:**
- They can operate over a very large range of temperatures.
- They are highly sensitive to changes in temperature.
- They are very accurate.
- The pressure changes linearly with temperature.

**Disadvantages:**
- They are cumbersome to operate. 
- They have a slow response.
- Use requires a skilled operator. 
- There is no direct temperature to voltage conversion.

---
### Radiation Pyrometry

Our final way of measuring temperature is via pyrometry, which is the technique we used to measure extremely high temperatures where it is not physically possible to get a thermometer into contact with the body.

This technique effectively uses the light emitted by the hot body to determine its temperature, for example an optical pyrometer will focus the light emitted by the glowing body onto a photodiode, which then produces a voltage proportional to the temperature of the object. 

A more analogue example of a pyrometer is that of the disappearing filament. Effectively, as we increase the current flowing through a filament of wire, we increase its temperature and cause it to glow due to resistive heating. The higher the current, the more brightly the filament will glow! If we then place this filament in front of the object whose temperature we would like to gauge, we can adjust the current until the glow of the filament matches that of the hot object and they become indistinguishable from one another. The current flowing through the wire is then a measure of the temperature of the object. 

The advantage of using pyrometers is that as they are non-contact, they do not change the temperature of the thing that you are measuring, they can be operated safely as they do not need to be directly exposed to high temperatures which may damage them and they respond to changes in temperature relatively quickly.   

---
## Questions

```{Warning}
**Atempt the questions before looking at the solutions otherwise you will not learn!** 


Students who look at mark schemes without going through the motions of a question often think "*yes, that's what I would have done*", but are then unable to perfom on exams as the knowledge was never consolidated through practise.
```

### Question 2

A $40\,$cm rod of unknown material, initially at $20$˚C, increases in temperature by $4.0$˚C and expands by $1.376 \times 10^{-5}\,$m. What material is the rod made from?

```{admonition} Q2 Solution
:class: dropdown

To determine what the material is, we must determine its thermal expansion coefficient. Thus we use the equation:

$ \alpha = \frac{\Delta L}{L_0 \Delta T}.$

Substituting in the known values gives:

$ \alpha = \frac{1.376 \times 10^{-5}}{0.40 \times 4},$

$ \alpha = 8.6 \times 10^{-6}\,$K^{-1}. 

Looking at the table in the notes, this is the thermal expansion coefficient of Titanium.

```
