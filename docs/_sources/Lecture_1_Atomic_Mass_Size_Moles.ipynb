## Building Blocks

Physics spans all scales of the universe, from the fundamental particles that make up everything around us to complex systems spanning the size of the observable universe. In order to describe these systems we first must have a solid understanding of the properties of matter. Let us begin by defining some terms.


### Atoms

An atom is an incredibly complex physical system comprising a series of *subatomic* particles. At this level we will look at a simplified model of the atom which comprises protons, neutrons and electrons. 

![](Figures/atom.png)

The protons and neutrons have an almost identical mass, but whereas the proton is positively charged, the neutron has no charge i.e. it is neutral. In an atom protons and neutrons are found within a small dense structure known as the nucleus. Orbiting the nucleus is are the electrons. The size of the atom is given by the orbit of the outer most electrons, which itself is approximately 10000 times larger than the size of the nucleus!

The number of protons in the nucleus dictates what *element* the atoms is. For example if an atom has 6 protons it is carbon, whereas if it has 8 it is oxygen. In a neutrally charged atom there are the same number of protons as electrons. The electron configuration of an atom dictate how it will interact with other atoms (e.g. for bonds etc). This is where physics becomes chemistry.

Chemical elements can be grouped according to their properties, which is how they are arranged within the *periodic table* as shown below, for example on the far left hand side are the reactive alkali metals, in the centre we find the transition metals and on the far right are the unreactive nobel gases.

![](Figures/PeriodicTable.png)

For each element, the periodic table will provde the *atomic number* of the element (e.g. number of protons), the element symbol, and the name of the element. More detailed periodic tables may also give masses and neutron numbers of the most stable isotopes. 

An isotope is simply an element with a different number of neutrons, for example, an atom of oxygen will always have 8 protons, but it may have between 3 and 20 neutrons. Not all isotopes are stable, but this is something that we will look at in the final unit "Atomic and Nuclear Physics." 

### Molecules

When individual atoms bond together they form a molecule. A molecule could be as simple as two hydrogen atoms in a diatomic system $H_2$, or be made of thousands of atoms like a a molecule of DNA. In theory there are an unlimited number of molecules that can be made.

Modelled here is a molecule of ATP, adenosine triphosphate, this molecule is synthesised biologically during cellular respiration and may then be split up using a molecule of water (hydrolysis) into a molecule of ADP and a phosphate group. This in turn releases energy that can be used in other metabolic processes. An average person actually synthesises and hydrolyses around $50\,kg$ of ATP per day!

![](Figures/atp.png)

```{Warning}
Ensure that you understand the difference between a nucleus, an atom and a molecule is. Often students use these terms interchangably despite the fact that they describe very different things. Doing so will result in lost marks in an assignment or exam.
```

## Atomic Mass Unit

When we are considering systems at this scale we need to use units that are appropriate. For example, the SI unit for mass is the kilogram, however an electron has a mass of approximately $9.11 \times 10^{-31}\, kg$. To put this in words, an electron has a mass of is approximately a thousand-billion-billion-billionth of a kilogram ($0.000000000000000000000000000000911\,kg$). This is not an appropriate unit to use... 

A better unit to use is the *atomic mass unit* (AMU, amu, u) often referred to as the *Dalton* (Da). One atomic mass unit is defined as *one-twelth the mass of a neutral unbound carbon-12 atom in its nuclear and electronic ground states at rest*. 

The conversion between amu and kg is: $1 \, amu = 1.660538921(73) \times 10^{-27} \, kg$. The number in brakets at the end represents the uncertainty associated with that value and may also be written as $(1.660538921 \pm 0.000000073) \times 10^{-27} \, kg$.

The proton and neutron have masses that are close to, but not exactly, equal to one atomic mass unit. One reason for this is that despite protons and neutrons both being nucleons, they themselves are made of fundamental particles which results in them having ever so slightly different masses.

## Moles

When we are considering how many units (things - atoms, molecules, etc.) make up an amount of material that fits the scale we experience (e.g. grams and kilos) it is not convinent to specifcy that there are a trillion-billion of them. Instead we use the unit the *mole*.

The mole is the SI unit of measurement for the amount of a substance. It is defined as exactly $6.02214076 \times 10^{23}$. This can be a number of atoms, molecules, or general units of stuff. The number above is known as Avogadro’s number, $N_A$ and it has units of mol$^{-1}$. 

Avogadro's number is defined as *the number of atoms in $12\,g$ of carbon-12*. This means that by virtue of how the atomic mass unit and Avogadro's number are defined, one can determine the value of either if the other is known.

By definition $12$ atomic mass units are equal to the mass of one carbon-12 atom, and $12\, g$ of carbon is made up of $N_A$ atoms of carbon-12. This can be expressed mathermatically as:

$12 \, u \times N_A = 12 \, g,$

thus

$ \therefore 1\,u \times N_A = 1 \, g,$

$ \therefore 1\,u = \frac{1}{N_A} \, g,$ 

$ \therefore 1\,u = \frac{1}{6.02214076 \times 10^{23}} \, g,$ 

$ \therefore 1\,u = 1.6605391 \times 10^{-24} g,$

$ \therefore 1\,u = 1.6605391 \times 10^{-27} kg.$

Although both values are provided on your constants sheet, knowing this formula may help when sanity checking work.

```{adominition} Important 

If I have $6.02214076 x 10^{23}$ atoms of oxygen, I have 1 mole of oxygen. 

If I have an Avagadro's number of anything, I have a mole of that thing. 

A mole is simply $6.02214076 x 10^{23}$ of anything!

$1\, u = \frac{1}{N_A \times 1000}$
```

### Molar Mass

The molar mass is just the mass of one mole of a substance. The unit of molar mass is typically $g \,mol^{-1}$, however to perform calculations it is usually best to convert this into SI units (e.g. kg mol-1). To convert between mass, molar mass and number of moles we use the relationship:

$ \text{Mass} = \text{Number of Moles} \times \text{Molar Mass} $

Avogadro's number was chosen so that the mass of one mole of a compound (g) is numerically equal to the average mass of one molecule of the compound in atomic mass units. For example, one mole of water contains $6.02214076 x 10^{23}$ molecules of water, and has a total mass of ~$18.015\,g$. The mean mass of one water molecule is $18.015 \, amu$.

### Molar Volume

Similar to the molar mass, we also have a term that relates the number of moles of a substance to the amount of volume it occupies. This is known as the *molar volume*, which itself has units of $m^3 \,mol^{-1}$. For standard temperature and pressure, one mole of an ideal gas will occupy $0.0224\, m^3$ of volume. Note: we will look at what an *ideal gas* is in the next lecture.

---
## Relative Atomic Mass

The relative atomic mass, $A_r$, of an atom/molecule can be considered as the mass of the atom/molecule with respect to the atomic mass unit. In this example, the atomic mass unit represents the scale for comparing the masses of different atoms/molecules with. The relative atomic mass is dimensionless:

$ A_r = \frac{\text{Mass of Atom (kg)}}{1u}.$

The exact same principle applies for the relative molecular mass:

$ M_r = \frac{\text{Mass of Molecule (kg)}}{1u}.$

For example, Lithium has a molar mass of $6.94 \, g \, mol^{-1}$, thereofre its relative atomic mass is $A_r = 6.94$. Hydrogen has a relative atomic mass $A_r$ of 1.00, thereofre its molar mass is $1 \, g \, mol^{-1}$.

Often relative molecular masses are rounded to the nearest integer. The molar mass of a substance is simply the sum of the molar masses of its consistent atoms, which can be obtained directly from the relative atomic masses from a periodic table.

```{admonition} Important

You may be asked to calculate approximate molecular masses for simple molecules in exams. You will be expected to know the following atomic values:

|Element|Symbol|Molar Mass ($g \, mol^{-1}$)|
|---|---|---|
| Carbon | C | 12 (by definition) |
| Hydrogen | H | 1 (approximate) |
|Helium | He | 4 (approximate) |
| Oxygen | O | 16 (appriximate) |

**You will not be provided witha  period table of the elements in a physics exam!**

```

## Density

Density, $\rho$, is simply defined as the amount of *mass per unit volume*, which can be expressed mathematically as:

$ \rho = \frac{M}{V}. $

The unit for density is therefore $kg\, m^{-3}$. It is expected that you will be able to calculate the densities of solids, liquids and gases as part of this unit.

## Determination of Molecular Sizes

The “oil-film experiment” can be used to estimate the molecular sizes of oils or combinations of immiscible liquids (liquids that do not mix). An oil drop with a known volume is dropped into a water tray. The oil drop will spread out and form a film on top of the water until the film is only one molecule thick and can spread out no further. Making the following assumptions:

- the drop of oil was spherical with a diameter, $d$,
- the oil film is only one molecule thick and circular with a diameter, $D$,
- the volume of the patch equals the volume of the droplet,

we are able to determine the approximate size of the oil molecules. We assume that the thickness, $t$ of the oil film is approximately the size of the oil molecule. The volume of the film is given by:

$ V_f = \pi r^2 t, $

$ V_f = \pi \left(\frac{D}{2}\right)^2 t., $

The volume of the oil droplet is given by:

$ V_d = \frac{4}{3} \pi r^3, $ 

$ V_d =\frac{4}{3} \pi \left(\frac{d}{2}\right)^3. $

As the volume of oil remains constant regardless of whether it is is in a droplet shape or a circular film, the two volumes can be equated:

$ \pi \left(\frac{D}{2}\right)^2 t = \frac{4}{3} \pi \left(\frac{d}{2}\right)^3. $

This expression can be rearranged to for the thickness of the film and therefore approximate molecular size:

$ \pi \left(\frac{D}{2}\right)^2 t = \frac{4}{3} \pi \left(\frac{d}{2}\right)^3, $

$ \left(\frac{D}{2}\right)^2 t = \frac{4}{3} \left(\frac{d}{2}\right)^3, $

$ \frac{D^2} t = \frac{16}{3}\frac{d^3}{8}, $

$ D^2 t = \frac{16}{24} d^3,$

$ t = \frac{2}{3} \frac{d^3}{D^2}.$

### Question 2

The molecules forming an oil film are determined to have a size of approximately $0.60 \times 10^{-10}\,m$ using the oil drop method. If the radius of the drop of oil was $0.15\, mm$, determine the radius of the circular oil film formed on the surface of the water.

```{admonition} Q2 Solution

Recalling that the thickness of the oil drop (and approximate molecular size) can be determined by equating the volume of the drop to the volume of the film:

$ t = \frac{2}{3} \frac{d^3}{D^2},$

and rearranging for the diameter of the film, $D$, and then radius:

$ D^2 = \frac{2}{3} \frac{d^3}{t}, $

$ D = \sqrt{\frac{2}{3} \frac{d^3}{t}}, $

$ r = \frac{1}{2} \sqrt{\frac{2}{3} \frac{d^3}{t}}, $

substituting in the known values gives:

$ r = \frac{1}{2} \sqrt{\frac{2}{3} \frac{\left(2\times 0.15 \times 10^{-3}\right)^3}{0.6\times 10^{-10}}}, $

$ r = 0.097 \, m. $

```
