### Atoms

An atom is an incredibly complex physical system comprising a series of **subatomic** particles. At this level we will look at a simplified model of the atom which comprises protons, neutrons and electrons.

The protons and neutrons have an almost identical mass, but whereas the proton is positively charged, the neutron has no charge i.e. it is neutral. In an atom protons and neutrons are found within a small dense structure known as the nucleus. Orbiting the nucleus is are the electrons. The size of the atom is given by the orbit of the outer most electrons, which itself is approximately 10000 times larger than the size of the nucleus!

The number of protons in the nucleus dictates what **element** the atoms is. For example if an atom has 6 protons it is carbon, whereas if it has 8 it is oxygen. In a neutrally charged atom there are the same number of protons as electrons. The electron configuration of an atom dictate how it will interact with other atoms (e.g. for bonds etc). This is where physics becomes chemistry.

Chemical elements can be grouped according to their properties, which is how they are arranged within the **periodic table** as shown below, for example on the far left hand side are the reactive alkali metals, in the centre we find the transition metals and on the far right are the unreactive nobel gases.

```{figure} Figures/PeriodicTable.png
:width: 100%
```

For each element, the periodic table will provde the **atomic number** of the element (e.g. number of protons), the element symbol, and the name of the element. More detailed periodic tables may also give masses and neutron numbers of the most stable isotopes. 

An **isotope** is simply an element with a different number of neutrons, for example, an atom of oxygen will always have 8 protons, but it may have between 3 and 20 neutrons. Not all isotopes are stable, but this is something that we will look at in the final unit "Atomic and Nuclear Physics." 

An **ion** is an atom that has missing or extra electrons, giving it a net positive or negative charge. **Cations** (pronounced "cat ions") are positively charged ions as they have lost electrons whereas **anions** (pronounced "an ions") are atoms that have gained electrons and therefore are negatively charged ions.

 

### Molecules

When individual atoms bond together they form a molecule. A molecule could be as simple as two hydrogen atoms in a diatomic system $\text{H}_2$, or be made of thousands of atoms like a a molecule of DNA. In theory there are an unlimited number of molecules that can be made.

Modelled here is a molecule of ATP, adenosine triphosphate, that is synthesised biologically during cellular respiration and may then be split up using a molecule of water (hydrolysis) into a molecule of ADP and a phosphate group. This in turn releases energy that can be used in other metabolic processes. An average person actually synthesises and hydrolyses around $50\text{ kg}$ of ATP per day!

## Atomic Mass Unit

When we are considering systems at this scale we need to use units that are appropriate. For example, the SI unit for mass is the kilogram, however an electron has a mass of approximately $9.11 \times 10^{-31}\, kg$. To put this in words, an electron has a mass of is approximately a thousand-billion-billion-billionth of a kilogram $(0.000000000000000000000000000000911\text{ kg})$. This is not an appropriate unit to use... 

A better unit to use is the **atomic mass unit** ("AMU", "amu", or more commonly "u") often referred to as the **Dalton** (Da). One atomic mass unit is defined as **one-twelth the mass of a neutral unbound carbon-12 atom in its nuclear and electronic ground states at rest**. 

The conversion between amu and kg is: $1 \text{ amu} = 1.660538921(73) \times 10^{-27} \text{ kg}$. The number in brakets at the end represents the uncertainty associated with that value and may also be written as $(1.660538921 \pm 0.000000073) \times 10^{-27} \text{ kg}$.

The proton and neutron have masses that are close to, but not exactly, equal to one atomic mass unit. One reason for this is that despite protons and neutrons both being nucleons, they themselves are made of fundamental particles which results in them having ever so slightly different masses.

## Moles

When we are considering how many units (things - atoms, molecules, etc.) make up an amount of material that fits the scale we experience (e.g. grams and kilos) it is not convinent to specify that there are a trillion-billion of them. Instead we use the unit the **mole** with unit symbol "mol".

The mole is the SI unit of measurement for the amount of a substance. It is defined as **exactly** $6.02214076 \times 10^{23}$. This can be a number of atoms, molecules, or general units of stuff. The number above is known as Avogadro’s number, $N_A$ and it has units of mol$^{-1}$. 

Avogadro's number is defined as **the number of atoms in $12\text{ g}$ of carbon-12**. This means that by virtue of how the atomic mass unit and Avogadro's number are defined, one can determine the value of either if the other is known.

By definition $12$ atomic mass units are equal to the mass of one carbon-12 atom, and $12\text{ g}$ of carbon is made up of $N_A$ atoms of carbon-12. This can be expressed mathematically as:

\begin{align*}
12 \text{ u} \times N_A = 12 \text{ g}
\end{align*}

thus

\begin{align*}
1\text{ u} \times N_A &= 1 \text{ g}\\
\therefore 1\text{ u} &= \frac{1}{N_A} \text{ g}\\ 
    &= \frac{1}{6.02214076 \times 10^{23}} \text{ g}\\
    &= 1.6605391 \times 10^{-24} \text{ g}\\
\therefore 1\text{ u} &= 1.6605391 \times 10^{-27} \text{ kg}
\end{align*}

Although both values are provided on your constants sheet, knowing this formula may help when sense checking work.

```{admonition} Important 

If I have $6.02214076 \times 10^{23}$ atoms of oxygen, I have 1 mole of oxygen. Similarly if I have $6.02214076 \times 10^{23}$ atom of hydrogen I have 1 mole of hydrogen. The same **number** of atoms, even though the total **mass** is different.

If I have an Avagadro's number of anything, I have a mole of that thing. A mole is simply $6.02214076 \times 10^{23}$ of anything!
```

### Molar Mass

The molar mass is just the mass of one mole of a substance. The unit of molar mass is typically $g \text{ mol}^{-1}$, however to perform calculations it is usually best to convert this into SI units (e.g. $\text{kg mol}^{-1}$). To convert between mass, molar mass and number of moles we use the relationship:

\begin{align*}
\text{Mass} = \text{Number of Moles} \times \text{Molar Mass}
\end{align*}

Avogadro's number was chosen so that the mass of one mole of a compound (in grammes) is numerically equal to the average mass of one molecule of the compound in atomic mass units. For example, one mole of water contains $6.02214076 x 10^{23}$ molecules of water, and has a total mass of ~$18.015\text{ g}$. The mean mass of one water molecule is $18.015 \text{ amu}$.

### Molar Volume

Similar to the molar mass, we also have a term that relates the number of moles of a substance to the amount of volume it occupies. This is known as the **molar volume**, which itself has units of $m^3 \,mol^{-1}$. For standard temperature and pressure, one mole of an ideal gas will occupy $0.0224\, m^3$ of volume.  
Note: we will look at what an **ideal gas** is in the [next topic](TPS_Topic2).

## Relative Atomic Mass

The relative atomic mass, $A_r$, of an atom/molecule can be considered as the mass of the atom/molecule with respect to the atomic mass unit. In this example, the atomic mass unit represents the scale for comparing the masses of different atoms/molecules with. The relative atomic mass is dimensionless:

\begin{align*}
A_r = \frac{\text{Mass of Atom (kg)}}{1u}
\end{align*}

The exact same principle applies for the relative molecular mass:

\begin{align*}
M_r = \frac{\text{Mass of Molecule (kg)}}{1u}
\end{align*}

For example, Lithium has a molar mass of $6.94 \text{ g mol}^{-1}$, therefore its relative atomic mass is $A_r = 6.94$. Hydrogen has a relative atomic mass $A_r$ of 1.00, therefore its molar mass is $1 \text{ g mol}^{-1}$.

Often relative molecular masses are rounded to the nearest integer. The molar mass of a substance is simply the sum of the molar masses of its consistent atoms, which can be obtained directly from the relative atomic masses from a periodic table.

## Density

Density, $\rho$, is simply defined as the amount of **mass per unit volume**, which can be expressed mathematically as:

\begin{align*}
\rho = \frac{M}{V}
\end{align*}

The unit for density is therefore $\text{kg m}^{-3}$. It is expected that you will be able to calculate the densities of solids, liquids and gases as part of this unit.

## Determination of Molecular Sizes

The “oil-film experiment” can be used to estimate the molecular sizes of oils or combinations of immiscible liquids (liquids that do not mix). An oil drop with a known volume is dropped into a water tray. The oil drop will spread out and form a film on top of the water until the film is only one molecule thick and can spread out no further without creating holes or gaps in the film. Making the following assumptions:

- the drop of oil was spherical with a diameter, $d$,
- the oil film is only one molecule thick and is circular disc with a diameter, $D$,
- the volume of the patch equals the volume of the droplet (i.e. no oil mass is lost),

we are able to determine the approximate size of the oil molecules by assuming that the thickness, $t$, of the oil film is approximately the size of the oil molecule. The volume of the film disc is given by:

\begin{align*}
V_f &= \pi r^2 t\\
    &= \pi \left(\frac{D}{2}\right)^2 t
\end{align*}

The volume of the spherical oil droplet is given by:

\begin{align*}
V_d &= \frac{4}{3} \pi r^3\\ 
    &=\frac{4}{3} \pi \left(\frac{d}{2}\right)^3
\end{align*}

As the volume of oil remains constant regardless of whether it is is in a droplet shape or a circular film, the two volumes can be equated:

\begin{align*}
V_f &= V_d\\
\therefore \pi \left(\frac{D}{2}\right)^2 t &= \frac{4}{3} \pi \left(\frac{d}{2}\right)^3
\end{align*}

This expression can be rearranged to for the thickness of the film and therefore approximate molecular size:

\begin{align*}
\pi \left(\frac{D}{2}\right)^2 t &= \frac{4}{3} \pi \left(\frac{d}{2}\right)^3\\
\left(\frac{D}{2}\right)^2 t &= \frac{4}{3} \left(\frac{d}{2}\right)^3\\
\frac{D^2} t &= \frac{16}{3}\frac{d^3}{8}\\
D^2 t &= \frac{16}{24} d^3\\
\therefore t &= \frac{2}{3} \frac{d^3}{D^2}
\end{align*}

`````{admonition} Question 1.1 - Calculating molar masses
Calculate the molar masses for the following molecules: diatomic nitrogen ($\text{N}_2$), methane ($\text{CH}_4$) and hydrogen peroxide ($\text{H}_2\text{O}_2$).

First find the **molecular mass** before then calculating the **molar mass**.

```{tab-item} Solution
Diatomic nitrogen (nitrogen gas) is made from two nitrogen atoms, therefore the relative molecular mass is:

\begin{align*}
14 + 14 = 28
\end{align*}

therefore the molar mass of diatomic nitrogen is $28 \text{ g mol}^{-1}$. 

---
Methane is made from one carbon atom and four hydrogen atoms, therefore the relative molecular mass is:

\begin{align*}
12 + 4(1) = 16
\end{align*}

therefore the molar mass of methane is $16 \text{ g mol}^{-1}$. 

---

Hydrogen peroxide contains two hydrogen atoms and two oxygen atoms, therefore the relative molecular mas is:

\begin{align*}
2(1) + 2(16) = 34
\end{align*}

therefore the molar mass of hydrogen peroxide is $34 \text{ g mol}^{-1}$.
`````

`````{admonition} Question 1.3 - Oil drop experiment
````{tab-set}
```{tab-item} Question
The molecules forming an oil film are determined to have a size of approximately $0.60 \times 10^{-10}\text{ m}$ using the oil drop method. If the radius of the drop of oil was $0.15\text{ mm}$, determine the radius of the circular oil film formed on the surface of the water.
```

```{tab-item} Hint
This is a reworking of the [oil drop calculation](L1_OilDrop) shown previously.  

Be careful of the different units used in the question, and that the question is specifically using and requesting **radii** rather than diameters.
```

```{tab-item} Solution
Recalling that the thickness of the oil drop (and approximate molecular size) can be determined by equating the volume of the drop to the volume of the film:

\begin{align*}
t = \frac{2}{3} \frac{d^3}{D^2}
\end{align*}

and rearranging for the diameter of the film, $D$, and then radius:

\begin{align*}
D^2 &= \frac{2}{3} \frac{d^3}{t}\\
D &= \sqrt{\frac{2}{3} \frac{d^3}{t}}\\
r &= \frac{1}{2} \sqrt{\frac{2}{3} \frac{d^3}{t}}
\end{align*}

substituting in the known values gives:

\begin{align*}
r &= \frac{1}{2} \sqrt{\frac{2}{3} \frac{\left(2\times 0.15 \times 10^{-3}\right)^3}{0.6\times 10^{-10}}}\\
r &= 0.097 \, m
\end{align*}
`````
